## 1) Transforma√ß√µes ortogonais e a propriedade fundamental

**Defini√ß√£o (O(3))**

**Propriedade fundamental:** ortogonalidade  $ \iff $ preserva produto interno e norma.

$(Ru)\cdot(Rv) = u^T R^T R v$.

Reciprocamente, se $R^T R=I$, ent√£o $(Ru)\cdot(Rv)=u\cdot v$, logo $\|Ru\|=\|u\|$.

> **Grupo de rota√ß√µes:**

## 2) Determinante e interpreta√ß√£o geom√©trica

## 4) Rota√ß√µes infinitesimais, anti-simetria e forma geral
## 4.1 Deriva√ß√£o da antissimetria a partir da ortogonalidade

## 6) Comutador de rota√ß√µes e rela√ß√µes de comuta√ß√£o

Para dois geradores infinitesimais $A_i=[\hat e_i]_\times$ ( $i=x,y,z$ ):

**Demonstra√ß√£o (vetorial):**

**Forma matricial:**

**(N√£o) comutatividade finita:**

Se $i\neq j$, o termo $[A_i,A_j]\neq0$ ‚Üí **n√£o comutam**.  
Se $i=j$, **comutam**.

## 7.x Deriva√ß√£o detalhada dos geradores de rota√ß√£o qu√¢ntica $ \mathbf J $

---

### (1) Suposi√ß√£o de partida: representa√ß√£o unit√°ria

Assumimos um mapeamento
$$

\;\equiv\;
\hat I + \theta_i\,\hat A_i + \mathcal O(\theta^2),

$$
\boxed{\;\hat A_i \equiv \left.\frac{\partial \hat U}{\partial \theta_i}\right|_{\boldsymbol\theta=0}\;}
\quad\text{(sem definir se √© hermitiano).}

$$
\boxed{\;\hat A_i + \hat A_i^\dagger = 0\;} \;\Rightarrow\; \hat A_i^\dagger = -\hat A_i.

√â conveniente trabalhar com observ√°veis hermitianos. Definimos:
$$

\;}
$$

---

### (6) Forma **finita** (exponencial)

$$

### (7) Coment√°rio: representa√ß√µes projetivas e $SU(2)$

---

**Conclus√£o:** a partir apenas da unitariedade e da expans√£o perto da identidade, derivamos
