## 1) Transformações ortogonais e a propriedade fundamental

**Definição (O(3))**  
Uma transformação linear $ R \in \mathbb{R}^{3\times3} $ é **ortogonal** se  

$$
R^T R = I.
$$

**Propriedade fundamental:** ortogonalidade  $ \iff $ preserva produto interno e norma.  
Para vetores $ u,v \in \mathbb{R}^3 $:

$$
(Ru)\cdot(Rv) = u\cdot v \quad \Longleftrightarrow \quad R^T R = I.
$$

**Demonstração:**  
$(Ru)\cdot(Rv) = u^T R^T R v$.  
Se vale para todo $u,v$, então $R^T R = I$.  
Reciprocamente, se $R^T R=I$, então $(Ru)\cdot(Rv)=u\cdot v$, logo $\|Ru\|=\|u\|$.  
Portanto, transformações ortogonais preservam ângulos e comprimentos.

> **Grupo de rotações:**  
> $$ SO(3)=\{R\in O(3)\mid \det R=+1\}. $$



---

## 2) Determinante e interpretação geométrica

Da ortogonalidade:  
$$
\det(R^T R)=\det I=1 \Rightarrow (\det R)^2=1 \Rightarrow \det R=\pm1.
$$

- $ \det R=+1 $: **rotação própria** (preserva orientação)  
- $ \det R=-1 $: **isometria imprópria** (rotação + reflexão; inverte orientação)

**Interpretação:**  
$|\det R|$ mede o fator de volume.  
Como $|\det R|=1$, todas as ortogonais preservam volumes.  
O sinal distingue preservação (+) ou inversão (−) da orientação do triedro $(\hat x,\hat y,\hat z)$.

---

## 3) Matrizes de rotação 3D e (não) comutatividade

**Rotações em torno dos eixos cartesianos** (ângulo $ \theta $):

$$
R_x(\theta)=
\begin{pmatrix}
1&0&0\\
0&\cos\theta&-\sin\theta\\
0&\sin\theta&\cos\theta
\end{pmatrix},\quad
R_y(\theta)=
\begin{pmatrix}
\cos\theta&0&\sin\theta\\
0&1&0\\
-\sin\theta&0&\cos\theta
\end{pmatrix},\quad
R_z(\theta)=
\begin{pmatrix}
\cos\theta&-\sin\theta&0\\
\sin\theta&\cos\theta&0\\
0&0&1
\end{pmatrix}.
$$

Cada uma satisfaz $R_a^T R_a=I$ e $\det R_a=+1$.

**Não comutatividade:**  
Em geral, $R_a(\alpha)R_b(\beta)\neq R_b(\beta)R_a(\alpha)$ para $a\neq b$.  
Exceções:
- quando $a=b$ (mesmo eixo) → **comutam**: $R_a(\alpha)R_a(\beta)=R_a(\alpha+\beta)$;
- ou ângulos especiais que se “cancelam” por simetria.

---

## 4) Rotações infinitesimais, anti-simetria e forma geral
## 4.1 Derivação da antissimetria a partir da ortogonalidade

Começamos da condição fundamental de ortogonalidade:

$$
R^T R = I.
$$

Para uma rotação **infinitesimal**, escrevemos:

$$
R = I + \epsilon A, \quad \text{com } \epsilon \ll 1.
$$

Substituindo em $ R^T R = I $:

$$
(I + \epsilon A^T)(I + \epsilon A) = I + \epsilon(A^T + A) + \mathcal{O}(\epsilon^2).
$$

Para que isso seja válido até primeira ordem, devemos ter:

$$
A^T + A = 0 \quad \Rightarrow \quad A^T = -A.
$$

Logo, $A$ é **antissimétrica**, e os elementos diagonais satisfazem $A_{ii} = 0$.

---

## 4.2 Forma geral do gerador infinitesimal

A matriz $A$ pode ser escrita de forma geral como:

$$
A =
\begin{pmatrix}
0 & -a_3 & a_2 \\
a_3 & 0 & -a_1 \\
-a_2 & a_1 & 0
\end{pmatrix}
= \sum_{i=1}^{3} a_i J_i,
$$

onde $a_1,a_2,a_3$ são os **parâmetros infinitesimais de rotação** em torno dos eixos $x,y,z$,
e $J_i$ são os **geradores básicos**:

$$
J_1 =
\begin{pmatrix}
0 & 0 & 0 \\
0 & 0 & -1 \\
0 & 1 & 0
\end{pmatrix}, \quad
J_2 =
\begin{pmatrix}
0 & 0 & 1 \\
0 & 0 & 0 \\
-1 & 0 & 0
\end{pmatrix}, \quad
J_3 =
\begin{pmatrix}
0 & -1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 0
\end{pmatrix}.
$$

Essas matrizes representam **rotações infinitesimais em torno de $x, y, z$** respectivamente.

---

## 4.3 Propriedade fundamental dos geradores

Em notação de índices:

$$
(J_i)_{jk} = -\varepsilon_{ijk}.
$$

Logo, os $J_i$ satisfazem a álgebra de comutação:

$$
[J_i, J_j] = \varepsilon_{ijk} J_k.
$$

Esses são os **geradores da álgebra de Lie so(3)**.

---

## 4.4 Forma geral de uma rotação infinitesimal

Uma rotação infinitesimal em torno de um eixo unitário $\hat{n}$ de ângulo $\theta$ é:

$$
R(\hat{n}, \theta) = I + \theta\,(\hat{n}\cdot \mathbf{J}) + \mathcal{O}(\theta^2),
$$

onde

$$
\hat{n}\cdot \mathbf{J} = n_x J_1 + n_y J_2 + n_z J_3.
$$

---

## 4.5 Da forma infinitesimal à forma finita

Somando infinitas rotações pequenas (limite contínuo):

$$
R(\hat{n}, \theta)
= \lim_{N\to\infty}\left(I + \frac{\theta}{N}\,\hat{n}\cdot\mathbf{J}\right)^N
= e^{\,\theta\,\hat{n}\cdot\mathbf{J}}.
$$

Esta é a **forma exponencial das rotações finitas**.

> Assim, os $J_i$ são os **geradores infinitesimais** de rotações, e a exponencial deles produz as **rotações finitas**.
---

## 5) De infinitesimais a finitas: forma exponencial

Somando $N$ rotações infinitesimais $\delta\theta=\theta/N$:

$$
R(\hat n,\theta)
=\lim_{N\to\infty}\!\left(I+\frac{\theta}{N}A(\hat n)\right)^N
=\exp\!\big(\theta\,A(\hat n)\big).
$$

Em 3D, usando $A=[\hat n]_\times$ e $A^3=-A$, obtém-se a **fórmula de Rodrigues**:

$$
R(\hat n,\theta)=I+\sin\theta\,A+(1-\cos\theta)A^2.
$$

Essa forma reproduz $R_x,R_y,R_z$ quando $\hat n$ é um eixo cartesiano.

> **Analogia quântica:**  
> No espaço de Hilbert, $U(\hat n,\theta)=\exp(-\tfrac{i}{\hbar}\theta\,\hat n\cdot\mathbf J)$.  
> No espaço vetorial 3D clássico, $R=\exp(\theta A)$.

---

## 6) Comutador de rotações e relações de comutação

Para dois geradores infinitesimais $A_i=[\hat e_i]_\times$ ( $i=x,y,z$ ):

$$
[A_i,A_j] = A_iA_j - A_jA_i = \varepsilon_{ijk}\,A_k.
$$

**Demonstração (vetorial):**  
Como $A_i v = \hat e_i\times v$,

$$
A_i(A_j v) = \hat e_i\times(\hat e_j\times v)
= \hat e_j(\hat e_i\cdot v) - v(\hat e_i\cdot\hat e_j).
$$

Trocando $i\leftrightarrow j$ e subtraindo:

$$
[A_i,A_j]v = (\hat e_i\times\hat e_j)\times v
= \varepsilon_{ijk}\,\hat e_k\times v
= \varepsilon_{ijk}A_k v.
$$

Como vale para todo $v$: $ [A_i,A_j]=\varepsilon_{ijk}A_k. $

**Forma matricial:**

$$
(A_i)_{\ell m}=-\varepsilon_{i\ell m}, \quad
[A_i,A_j]_{\ell n}=\varepsilon_{ijk}(A_k)_{\ell n}.
$$

**(Não) comutatividade finita:**  
Para ângulos finitos pequenos $\alpha,\beta$, pelo BCH:

$$
e^{\alpha A_i} e^{\beta A_j}
= e^{\alpha A_i + \beta A_j + \tfrac{1}{2}\alpha\beta[A_i,A_j]+\cdots}.
$$

Se $i\neq j$, o termo $[A_i,A_j]\neq0$ → **não comutam**.  
Se $i=j$, **comutam**.

---

### Identidades úteis

- $A=[\mathbf a]_\times$ com $A v=\mathbf a\times v$
- $A^2=\mathbf a\mathbf a^T-\|\mathbf a\|^2 I$
- Para $\|\hat n\|=1$: $A^3=-A$
- Levi-Civita: $\varepsilon_{ilm}\varepsilon_{jmn}=\delta_{ij}\delta_{ln}-\delta_{in}\delta_{lj}$
- BCH: $e^X e^Y = e^{X+Y+\tfrac12[X,Y]+\cdots}$

---

> **Resumo final:**  
> Rotações 3D são isometrias (ortogonais com $\det=+1$);  
> infinitesimalmente, são geradas por matrizes anti-simétricas;  
> composições finitas nascem da exponencial;  
> e a não comutatividade é regida por  
> $$[A_i,A_j]=\varepsilon_{ijk}A_k,$$  
> a álgebra $so(3)$.

## 7) Rotações quânticas: operadores unitários e geradores

**Wigner:** simetrias que preservam probabilidades são representadas por operadores unitários/antiunitários. Para rotações, tomamos **unitários**:
$$
U(R_2)\,U(R_1)=U(R_2R_1),\qquad U(I)=\mathbf 1,\qquad U(R)^{-1}=U(R)^\dagger .
$$

> Em MQ, representações podem ser **projetivas**: o grupo físico das rotações é $SO(3)$, mas a representação quântica fundamental é seu **revestimento duplo** $SU(2)$. Consequência: para spin $1/2$, $U(2\pi)=-\mathbf 1$ e $U(4\pi)=\mathbf 1$.

**Definição dos geradores (infinitesimais):**
se $R(\boldsymbol\theta)$ é uma rotação parametrizada por $\boldsymbol\theta\in\mathbb R^3$,
$$
U(\boldsymbol\theta)=\mathbf 1-\frac{i}{\hbar}\,\boldsymbol\theta\!\cdot\!\mathbf J+\mathcal O(\theta^2),\qquad
J_k\equiv i\hbar\left.\frac{\partial U(\boldsymbol\theta)}{\partial \theta_k}\right|_{\boldsymbol\theta=0}.
$$
Como $U$ é unitário, os geradores são **Hermitianos**: $J_k^\dagger=J_k$.

**Forma finita (exponencial):**
$$
\boxed{~U(\hat n,\theta)=\exp\!\left(-\frac{i}{\hbar}\,\theta\,\hat n\!\cdot\!\mathbf J\right).~}
$$

## 8) Álgebra de comutação dos $J_i$ via composição (BCH)

Considere duas rotações infinitesimais $\alpha$ e $\beta$ em torno de eixos $i$ e $j$:
$$
U_i(\alpha)=e^{-\frac{i}{\hbar}\alpha J_i},\qquad
U_j(\beta)=e^{-\frac{i}{\hbar}\beta J_j}.
$$
Pelo **Baker–Campbell–Hausdorff**,
$$
U_i(\alpha)U_j(\beta)=
\exp\!\left(-\frac{i}{\hbar}(\alpha J_i+\beta J_j)
+\frac{1}{2}\left(-\frac{i}{\hbar}\right)^2\alpha\beta[J_i,J_j]+\cdots\right).
$$
Mas compor rotações no espaço dá outra rotação (em torno de $k$) já na **segunda ordem**. Logo,
$$
[J_i,J_j]=i\hbar\,\varepsilon_{ijk}\,J_k.
$$
Essa é a álgebra de Lie **su(2)/so(3)** no espaço de Hilbert.
