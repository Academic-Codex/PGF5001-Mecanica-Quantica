# Aula 20 — Degenerate Perturbation Theory

A degenerescência de autovalores em $H_0$ impede que a teoria de perturbação
usual seja aplicada diretamente. Isso ocorre porque, se vários kets
têm a mesma energia $E_D^{(0)}$, o denominador 
$E_n^{(0)} - E_k^{(0)}$ da fórmula não-degenerada se anula.  
Logo, precisamos reorganizar o tratamento: primeiro "descobrimos"
quais combinações lineares dentro do subespaço degenerado realmente 
são os autovetores corretos quando a perturbação é ligada.

A intenção da aula é mostrar **como isolar a dinâmica no espaço degenerado**,
**como resolver a correção de primeira ordem de energia** mediante a diagonalização 
de uma matriz menor $g \times g$, e **como reconstruir os estados perturbados**.

---

## 1. Setup da Teoria de Perturbação

Começamos com o Hamiltoniano perturbado:

$$
H(\lambda) = H_0 + \lambda H_1,
\qquad \lambda \ll 1.
$$

Assumimos expansões analíticas:

### Expansão da energia
$$
E_n(\lambda) =
E_n^{(0)} + \lambda\,\Delta_n^{(1)} +
\lambda^2\,\Delta_n^{(2)}
+ \lambda^3\,\Delta_n^{(3)} + \cdots
$$

### Expansão do estado
$$
|n(\lambda)\rangle =
|n^{(0)}\rangle
+ \lambda |n^{(1)}\rangle
+ \lambda^2 |n^{(2)}\rangle + \cdots
$$

No caso não degenerado, as fórmulas para $\Delta_n^{(1)}$ e $\Delta_n^{(2)}$ são:

$$
\Delta_n^{(1)} = \langle n^{(0)}|H_1|n^{(0)}\rangle,
$$

$$
\Delta_n^{(2)} = \sum_{k\neq n}
\frac{|V_{nk}|^2}{E_n^{(0)} - E_k^{(0)}},
\qquad \text{com } V_{nk} = \langle n^{(0)}|H_1|k^{(0)}\rangle.
$$

Essas fórmulas **quebram completamente** quando $E_n^{(0)} = E_k^{(0)}$, pois o denominador zera. Por isso precisamos reformular a teoria.


### Correção de ordem 1
$$
\Delta_n^{(1)} 
= V_{nn}
= \langle n^{(0)}|H_1|n^{(0)}\rangle
$$

### Correção de ordem 2
$$
\Delta_n^{(2)} 
= \sum_{k\neq n}
\frac{|V_{nk}|^2}{E_n^{(0)} - E_k^{(0)}}
$$

### Estado de primeira ordem
$$
|n^{(1)}\rangle
= \sum_{k\neq n}
\frac{V_{kn}}{E_n^{(0)} - E_k^{(0)}}
|k^{(0)}\rangle
$$

Aqui já vemos o problema: se $E_n^{(0)} = E_k^{(0)}$, o denominador zera.  
Isso motiva o novo formalismo.

---

## 2. Estrutura de um Subespaço Degenerado

Assumimos que um certo autovalor $E_D^{(0)}$ de $H_0$ possui degenerescência $g$. Os estados degenerados são:

$$
H_0 |m^{(0)}\rangle = E_D^{(0)} |m^{(0)}\rangle,
\qquad m=1,\dots,g.
$$

Chamamos esse conjunto de **subespaço degenerado**.

Definimos dois projetores fundamentais:

### Projetor no subespaço degenerado:
$$
P_D = \sum_{m\in D} |m^{(0)}\rangle\langle m^{(0)}|.
$$

### Projetor no subespaço complementar:
$$
P_\perp = \sum_{k\notin D} |k^{(0)}\rangle\langle k^{(0)}|.
$$

Eles satisfazem:

$$
P_D + P_\perp = 1,
\qquad
P_D P_\perp = 0,
\qquad
[H_0, P_D] = [H_0, P_\perp] = 0.
$$

**Intenção pedagógica:**  
Separar o espaço de Hilbert em duas regiões:  
1. A parte onde ocorre a degenerescência (precisa de tratamento especial).  
2. A parte não degenerada, onde a perturbação funciona normalmente.

---

## 3. O Problema do Autovetor Perturbado

Escrevemos o autovetor perturbado como:

$$
|\ell\rangle = P_D|\ell\rangle + P_\perp|\ell\rangle
$$

e começamos da equação de autovalor:

$$
(H_0 + \lambda H_1)|\ell\rangle = E |\ell\rangle.
$$

Aplicamos $P_D$ e $P_\perp$ a essa equação e obtemos duas equações acopladas.

### Equação projetada em $D$:
$$
0 =
P_D(E - H_0 - \lambda H_1)P_D|\ell\rangle
+ P_D(E - H_0 - \lambda H_1)P_\perp|\ell\rangle.
$$

### Equação projetada em $\perp$:
$$
0 =
P_\perp(E - H_0 - \lambda H_1)P_D|\ell\rangle
+ P_\perp(E - H_0 - \lambda H_1)P_\perp|\ell\rangle.
$$

**Intenção pedagógica:**  
Essas duas equações são análogas a um sistema linear acoplado. Vamos resolver a segunda (mais simples) para $P_\perp|\ell\rangle$ e injetar o resultado na primeira. Esse procedimento elimina o subespaço complementar e deixa uma equação **somente dentro da degenerescência**, que será nossa equação efetiva.

---


## 4. Resolvendo o Complemento $P_\perp|\ell\rangle$

Da equação em $\perp$ isolamos:

$$
(E - H_0) P_\perp|\ell\rangle
=
\lambda\, P_\perp H_1 P_D|\ell\rangle
+ \mathcal{O}(\lambda^2).
$$

Como $H_0$ não é degenerado em $\perp$,
o operador $(E_D^{(0)} - H_0)$ é invertível nesse subespaço. Logo:

$$
P_\perp|\ell^{(1)}\rangle
=
\frac{1}{E_D^{(0)} - H_0}
P_\perp H_1 P_D |\ell^{(0)}\rangle.
$$

Em forma explícita:

$$
P_\perp|\ell^{(1)}\rangle
=
\sum_{k\notin D}
\frac{
\langle k^{(0)}|H_1|\ell^{(0)}\rangle
}{E_D^{(0)} - E_k^{(0)}}
|k^{(0)}\rangle.
$$

**Intenção:**  
O componente do autovetor fora da degenerescência é totalmente determinado pelo efeito de $H_1$ sobre o componente degenerado. Ele não contém liberdade: só o subespaço degenerado é “indeterminado”.

---

## 5. Inserindo em $P_D$ — A Equação Efetiva

Agora substituímos $P_\perp|\ell\rangle$ na equação projetada em $P_D$.

Mantendo apenas termos até ordem $\lambda$, obtemos:

$$
(E - E_D^{(0)}) P_D|\ell\rangle
=
\lambda\, P_D H_1 P_D \, P_D|\ell\rangle.
$$

Como $P_D|\ell\rangle \equiv |\ell^{(0)}\rangle$ está dentro da degenerescência:

### Equação efetiva dentro do espaço degenerado:
$$
P_D H_1 P_D |\ell^{(0)}\rangle
=
\Delta^{(1)}_\ell |\ell^{(0)}\rangle.
$$

Chamamos:

$$
\widetilde{H}_1 = P_D H_1 P_D.
$$

Temos assim:

$$
\widetilde{H}_1 |\ell^{(0)}\rangle
= \Delta^{(1)}_\ell |\ell^{(0)}\rangle.
$$

**Intenção pedagógica (mais importante da aula):**  
A primeira ordem da perturbação resolve a degenerescência por meio da diagonalização de $H_1$ restrito ao subespaço degenerado. O problema de dimensão arbitrária reduz-se a diagonalizar uma matriz $g\times g$.

---


## 6. A Matriz Efetiva no Subespaço Degenerado

A matriz reduzida tem elementos:

$$
(\widetilde{H}_1)_{mn}
=
\langle m^{(0)}| H_1 | n^{(0)}\rangle,
\qquad m,n = 1,\dots,g.
$$

A equação de autovalores:

$$
\det\left( \widetilde{H}_1 - \Delta^{(1)}_\ell I \right) = 0
$$

dá as correções de primeira ordem $\Delta_{\ell}^{(1)}$.

Os autovetores:

$$
|\ell^{(0)}\rangle = \sum_{m=1}^g c_m^{(\ell)} |m^{(0)}\rangle
$$

determinam **quais combinações lineares são os estados físicos verdadeiros**.

---

## 7. Correção de Segunda Ordem

Uma vez resolvida a degenerescência:

$$
\Delta^{(2)}_\ell
=
\sum_{k \notin D}
\frac{
|\langle k^{(0)}|H_1|\ell^{(0)}\rangle|^2
}{E_D^{(0)} - E_k^{(0)}}.
$$

Esse resultado é formalmente idêntico ao caso não-degenerado,
exceto que não se somam termos dentro do próprio subespaço degenerado
(pois esses já foram tratados).

**Intenção pedagógica:**  
Depois que a degenerescência é removida pela primeira ordem, tudo funciona como na teoria de perturbação usual.

---


## 8. Estado Perturbado Completo (até primeira ordem)

$$
|\ell\rangle
=
|\ell^{(0)}\rangle
+ \lambda
\sum_{k\notin D}
\frac{\langle k^{(0)}|H_1|\ell^{(0)}\rangle}{
E_D^{(0)} - E_k^{(0)}
}
|k^{(0)}\rangle
+ \cdots.
$$

---

## 9. Summary da Aula

1. Identifique o subespaço degenerado do Hamiltoniano $H_0$.  
2. Construa a matriz reduzida  
   $$
   (\widetilde{H}_1)_{mn}
   =
   \langle m^{(0)}|H_1|n^{(0)}\rangle.
   $$
3. Diagonalize essa matriz $g\times g$.  
4. Seus autovalores dão as correções de **primeira ordem** da energia.  
5. Seus autovetores definem os **verdadeiros estados de ordem zero**.  
6. As correções de segunda ordem são dadas pela fórmula usual,  
   excluindo contribuições dentro de $D$.  
7. Para ordens superiores, aplica-se a teoria usual com o cuidado de  
   nunca incluir projeções para dentro do subespaço degenerado.
