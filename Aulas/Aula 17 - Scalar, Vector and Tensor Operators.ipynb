### Resumo estruturado das notas de aula

---

# 1. Introdução: Por que estudar operadores sob rotações?

O objetivo desta aula é entender **como operadores quânticos se transformam quando aplicamos rotações**.  
Isso é fundamental na mecânica quântica porque:

- estados possuem simetrias,
- Hamiltonianos frequentemente são invariantes rotacionais,
- certos observáveis transformam-se como vetores (posição, momento, spin),
- outros como tensores (quadrupolo elétrico, tensores de spin),
- e alguns permanecem invariantes (energias, integrais escalares).

A ideia é classificar operadores segundo sua **transformação sob o grupo de rotações SO(3)**,
e depois aplicar essa classificação à construção de **tensores esféricos irreducíveis**, que são a base do formalismo de Wigner.

---

# 2. Operadores Escalares

## 2.1 Definição matemática

### Ideia da seção

Aqui queremos ligar duas maneiras de dizer que um operador é **escalar**:

1. **Definição geométrica:** ele é invariável sob qualquer rotação
   $$
   U(\mathcal R) \, K \, U^\dagger(\mathcal R) = K \quad \text{para toda rotação } \mathcal R.
   $$

2. **Definição algébrica:** ele **comuta com todos os componentes** do momento angular
   $$
   [K, J_i] = 0 \quad (i=x,y,z).
   $$

O objetivo é **derivar (2) a partir de (1)**.  
Para isso, olhamos a transformação para uma rotação infinitesimal e usamos uma expansão em série de Taylor do operador de rotação.

---

### Operador de rotação "normal"

Uma rotação finita de ângulo $\boldsymbol{\theta}$ em torno do eixo definido pelo versor $\hat{n}$ é gerada pelo operador de momento angular $\vec{J}$:

$$
U(\mathcal R) = U(\hat{n},\theta)
= \exp\!\left(-\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\, \theta\right).
$$

Aqui:
- $\vec{J}\cdot\hat{n} = J_x n_x + J_y n_y + J_z n_z$ é o componente de $\vec{J}$ na direção de $\hat{n}$;

---

### Por que expandir em série de Taylor?

A condição de escalar
$$
U(\mathcal R) K U^\dagger(\mathcal R) = K
$$
é verdadeira para **qualquer ângulo**.  

Se olharmos especificamente para uma rotação **infinitesimal** (ângulo muito pequeno $\theta$), a transformação fica próxima da identidade e podemos escrever $U$ como:

$$
U = 1 + (\text{algo})\times\theta + O(\theta^2).
$$

O “algo” que multiplica $\theta$ será exatamente o **gerador** da rotação.  
Ao multiplicar $U K U^\dagger$ e manter apenas os termos de primeira ordem em $\theta$, o coeficiente de $\theta$ vira um **comutador** com esse gerador.  
Impor que a transformação não mude $K$ para *todo* $\theta$ nos força a concluir que esse comutador é zero.

---

### Expansão de $U$ até primeira ordem

Expandimos a exponencial:

$$
U(\mathcal R)
= \exp\!\left(-\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
\approx 1 - \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta
+ O(\theta^2).
$$

O adjunto é

$$
U^\dagger(\mathcal R)
= \exp\!\left(+\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
\approx 1 + \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta
+ O(\theta^2),
$$

pois $\vec{J}$ é hermitiano.

---

### Aplicando a condição de escalar

A condição de invariância é
$$
U K U^\dagger = K.
$$

Substituindo as expansões:

$$
\begin{aligned}
U K U^\dagger
&\approx 
\left(1 - \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
K
\left(1 + \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right) \\
&=
K
- \frac{i}{\hbar}\,(\vec{J}\cdot\hat{n})\,\theta\, K
+ \frac{i}{\hbar}\,K\, (\vec{J}\cdot\hat{n})\,\theta
+ O(\theta^2).
\end{aligned}
$$

Agrupando os termos de primeira ordem em $\theta$:

$$
U K U^\dagger
\approx
K + \frac{i}{\hbar}\,\theta\,
\big(
K\,\vec{J}\cdot\hat{n}
- \vec{J}\cdot\hat{n}\, K
\big)
+ O(\theta^2).
$$

Ou seja,

$$
U K U^\dagger
\approx
K + \frac{i}{\hbar}\,\theta\, [K,\vec{J}\cdot\hat{n}] + O(\theta^2).
$$

Mas, pela definição de operador escalar, devemos ter

$$
U K U^\dagger = K
\quad \text{para qualquer } \theta.
$$

Isso só é possível se o coeficiente de $\theta$ for nulo:

$$
[K,\vec{J}\cdot\hat{n}] = 0.
$$

Como $\hat{n}$ é um versor **arbitrário**, essa igualdade implica que cada componente deve comutar separadamente:

$$
\boxed{
[K,J_x]=[K,J_y]=[K,J_z]=0
\quad \Rightarrow \quad
[K,J_i]=0 \;\; \forall i.}
$$

---

### Conclusão desta seção

> **Operadores escalares comutam com todos os componentes do momento angular**.  
> A expansão de $U$ para rotações infinitesimais é a ponte entre:
> - a definição geométrica $U K U^\dagger = K$ e  
> - a condição algébrica de comutação $[K,J_i]=0$.

## 3.1 Operadores vetoriais

### Ideia da seção

Aqui queremos fazer para **operadores vetoriais** o mesmo tipo de raciocínio que fizemos para os **escalares**:

1. Começamos com a **definição geométrica**: um operador vetorial deve transformar-se como um vetor 3D comum sob rotações;
2. A partir disso, extraímos a **condição algébrica** em termos de comutadores com o momento angular:
   $$
   [J_i, V_j] = i\hbar\,\varepsilon_{ijk} V_k.
   $$

De novo, o truque é olhar para **rotações infinitesimais** e expandir o operador de rotação $U$ em série de Taylor; a parte linear em $\theta$ é justamente o comutador com o gerador.

---

### Lei de transformação sob rotação finita

Por definição, um operador vetorial $ \vec{V} = (V_x,V_y,V_z) $ transforma-se como

$$
U(\mathcal R)\, V_i \, U^\dagger(\mathcal R)
= \sum_{j} R_{ij}(\mathcal R)\, V_j,
$$

onde $R_{ij}(\mathcal R)$ é a matriz de rotação usual em 3D (a mesma que age em vetores clássicos).

O que queremos é reescrever isso em termos de **comutadores com $J_i$**.

---

### Por que olhar rotações infinitesimais?

Assim como no caso escalar:

- Para uma rotação **finita**, a condição acima é complicada de manipular diretamente.
- Para uma rotação **infinitesimal**, podemos escrever
  $$
  U \approx 1 + (\text{algo})\times\theta, \qquad
  R_{ij} \approx \delta_{ij} + (\text{algo})\times\theta,
  $$
  e comparar termo a termo em $\theta$.

O “algo” no lado quântico será o **gerador $ \vec{J} $**, e no lado clássico será o **gerador da rotação na representação vetorial**, que envolve o símbolo de Levi-Civita $ \varepsilon_{ijk} $.  
Ao igualar os dois, descobrimos o comutador $[J_i, V_j]$.

---

### Operador de rotação infinitesimal

Para uma rotação de pequeno ângulo $\theta$ em torno de um versor $\hat{n}$, temos

$$
U(\hat{n},\theta)
= \exp\!\left(-\frac{i}{\hbar}\, \vec{J}\cdot \hat{n} \,\theta\right).
$$

Expandindo em série de Taylor até 1ª ordem em $\theta$:

$$
U(\hat{n},\theta)
\simeq 1 - \frac{i}{\hbar}\, (\vec{J}\cdot\hat{n})\,\theta
+ O(\theta^2),
$$

$$
U^\dagger(\hat{n},\theta)
\simeq 1 + \frac{i}{\hbar}\, (\vec{J}\cdot\hat{n})\,\theta
+ O(\theta^2),
$$

porque $ \vec{J} $ é hermitiano.

---

### Matriz de rotação infinitesimal (lado clássico)

Seja um vetor clássico $ \vec{a} $.  
Sob uma rotação de ângulo pequeno $\theta$ em torno de $\hat{n}$:

$$
\vec{a}' = \vec{a} + \theta\, (\hat{n} \times \vec{a}).
$$

Em componentes:

$$
a'_i
= a_i + \theta \, (\hat{n}\times\vec{a})_i
= a_i + \theta\, \varepsilon_{i\ell j} n_\ell a_j.
$$

Comparando com $ a'_i = R_{ij} a_j $, obtemos

$$
R_{ij}(\hat{n},\theta)
\simeq \delta_{ij} + \theta\, \varepsilon_{i\ell j}\, n_\ell
+ O(\theta^2).
$$

É essa forma de $R_{ij}$ que vamos usar no lado direito da transformação de $V_i$.

---

### Agora, equacionando os dois lados

Partimos da definição vetorial:

$$
U(\hat{n},\theta)\, V_i \, U^\dagger(\hat{n},\theta)
= \sum_j R_{ij}(\hat{n},\theta)\, V_j.
$$

#### Lado esquerdo — usando o $U$ expandido

Substituindo as expansões de $U$ e $U^\dagger$:

$$
\begin{aligned}
U V_i U^\dagger
&\simeq 
\left(1 - \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
V_i
\left(1 + \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right) \\
&=
V_i
- \frac{i}{\hbar}\,\theta\, (\vec{J}\cdot\hat{n}) V_i
+ \frac{i}{\hbar}\,\theta\, V_i (\vec{J}\cdot\hat{n})
+ O(\theta^2) \\
&=
V_i + \frac{i}{\hbar}\,\theta\,
\big[ \, \vec{J}\cdot\hat{n},\, V_i \,\big]
+ O(\theta^2).
\end{aligned}
$$

#### Lado direito — usando $R_{ij}$ expandido

$$
\begin{aligned}
\sum_j R_{ij} V_j
&\simeq
\sum_j \left(\delta_{ij} + \theta\, \varepsilon_{i\ell j}\, n_\ell\right) V_j \\
&=
V_i + \theta \sum_{j,\ell} \varepsilon_{i\ell j} n_\ell V_j
+ O(\theta^2).
\end{aligned}
$$

---

### Igualando termos de primeira ordem em $\theta$

A igualdade $ U V_i U^\dagger = \sum_j R_{ij} V_j $ deve valer para **todo** $\theta$, então os coeficientes de $\theta$ precisam coincidir:

$$
\frac{i}{\hbar}
\big[ \, \vec{J}\cdot\hat{n},\, V_i \,\big]
=
\sum_{j,\ell} \varepsilon_{i\ell j} n_\ell V_j.
$$

Escrevendo o comutador à esquerda com índices:

$$
\vec{J}\cdot\hat{n} = \sum_\ell J_\ell n_\ell
\quad\Rightarrow\quad
\big[ \, \vec{J}\cdot\hat{n},\, V_i \,\big]
= \sum_\ell n_\ell [J_\ell, V_i].
$$

Substituindo:

$$
\frac{i}{\hbar} \sum_\ell n_\ell [J_\ell, V_i]
=
\sum_{j,\ell} \varepsilon_{i\ell j} n_\ell V_j.
$$

Como $ \hat{n} $ é **arbitrário**, os coeficientes de $n_\ell$ devem ser iguais para cada $\ell$:

$$
\frac{i}{\hbar} [J_\ell, V_i]
= \sum_{j} \varepsilon_{i\ell j} V_j.
$$

---

### Colocando na forma padrão $[J_i,V_j] = i\hbar \varepsilon_{ijk} V_k$

Agora é só reorganizar índices (apenas renomeando letras mudas de somatório):

- troque $\ell \to i$,  
- troque $i \to j$,  
- troque $j \to k$.

Obtemos:

$$

\frac{i}{\hbar} [J_i, V_j]
= \sum_{k} \varepsilon_{j i k} V_k.
$$

Usando a propriedade antissimétrica de $\varepsilon_{jik} = -\varepsilon_{ijk}$ e absorvendo o sinal em $i$, temos:

$$
\boxed{
[J_i, V_j]
= i\hbar\, \varepsilon_{ijk} V_k.}
$$

---

### Conclusão desta seção

> Um operador vetorial é definido por transformar-se como um vetor clássico sob rotações:
> $$
> U(\mathcal R) V_i U^\dagger(\mathcal R) = \sum_j R_{ij} V_j.
> $$
>  
> Ao considerar rotações infinitesimais e expandir tanto $U$ quanto $R_{ij}$ em série de Taylor, obtemos a condição algébrica:
> $$
> [J_i, V_j] = i\hbar\,\varepsilon_{ijk} V_k,
> $$
> que mostra que os componentes de $\vec{V}$ formam uma representação vetorial (spin 1) do grupo de rotações SO(3).

## 4. Operadores Tensoriais

### Intenção conceitual

Operadores tensoriais são o passo natural após escalares (rank 0) e vetores (rank 1).  
Eles aparecem quando combinamos operadores vetoriais entre si (ex.: $r_i r_j$, $p_i p_j$, $L_i S_j$), ou quando lidamos com grandezas que possuem mais de uma direção associada: tensores de quadrupolo, tensores de deformação, derivadas segundas, etc.

A regra geral é:

> **Um operador tensorial de rank $n$ possui $n$ índices e transforma com $n$ matrizes de rotação**, exatamente como tensores clássicos.

Esta seção mostra:
1. a definição formal via rotações finitas,
2. como ela generaliza a transformação de escalares e vetores,
3. por que tensores surgem naturalmente na física.

---

## 4.1 Definição geral

### Motivação

Se um operador tiver **um índice**, como $V_i$, ele deve transformar como um vetor:

$$
U V_i U^\dagger = R_{ij} V_j.
$$

Se tiver **zero índices**, ele é escalar:

$$
UKU^\dagger = K.
$$

Então, para um operador com **dois índices**, a única transformação rotacional consistente é:

$$
T_{ij} \;\longrightarrow\; R_{ik} R_{j\ell} T_{k\ell}.
$$

Isso é exatamente a regra clássica para tensores 3D.  
Na MQ apenas “envolvemos” essa transformação entre $U$ e $U^\dagger$.

---

### ### Rank 2 (um caso fundamental)

Queremos que o operador $T_{ij}$ transforme da mesma forma que um tensor cartesiano clássico:

$$
U(\mathcal R)\, T_{ij}\, U^\dagger(\mathcal R)
= \sum_{k\ell} R_{ik}(\mathcal R)\, R_{j\ell}(\mathcal R)\, T_{k\ell}.
$$

**Interpretação:**
- O índice $i$ se transforma como o índice de um vetor,
- O índice $j$ também,
- Cada índice incorpora seu próprio fator da matriz de rotação.

Isso permite que $T_{ij}$ capture simultaneamente informações associadas a duas direções distintas.

---

### Rank arbitrário

Se o tensor tem $n$ índices:

$$
T_{i_1 i_2 \dots i_n},
$$

a rotação deve agir separadamente em cada índice:

$$
U T_{i_1 i_2 \dots i_n} U^\dagger
=
\sum_{j_1 \dots j_n}
R_{i_1 j_1}
R_{i_2 j_2}
\cdots
R_{i_n j_n}\,
T_{j_1 j_2 \dots j_n}.
$$

Esta é a **definição geral de operador tensorial em MQ**.

---

## 4.2 Origem dessa regra: porque dois fatores de $R$?

### Intuição geométrica

Se você pensa em um vetor clássico $v$, ele transforma assim:

$$
v'_i = R_{ij} v_j.
$$

Agora pense em um objeto com **duas direções associadas**, como:

- uma superfície com orientação definida nos eixos $i$ e $j$,
- um operador que relaciona duas direções espaciais (ex.: momento de inércia),
- um produto de vetores $V_i W_j$.

Esse objeto exige dois fatores da rotação, porque cada direção (cada índice) gira independentemente.

---

## 4.3 Derivação por infinitesimais (análoga ao caso vetorial)

Tal como fizemos para os vetores, podemos reproduzir a lei de transformação usando rotações infinitesimais.

### Usando o operador infinitesimal:

$$
U \approx 1 - \frac{i}{\hbar} (\vec{J}\cdot\hat{n})\,\theta.
$$

### Aplicando a transformação:

$$
U T_{ij} U^\dagger
\approx
T_{ij}
+ \frac{i}{\hbar}\,\theta\,
\left[ \vec{J}\cdot\hat{n},\, T_{ij} \right].
$$

### Usando a transformação clássica infinitesimal:

Um tensor clássico de rank 2 transforma-se como:

$$
T'_{ij}
=
T_{ij}
+ \theta\big[
(\hat{n}\times)_i{}^k\, T_{kj}
+ (\hat{n}\times)_j{}^\ell\, T_{i\ell}
\big].
$$

Onde:

$$
(\hat{n}\times)_i{}^k = \varepsilon_{i m k} n_m.
$$

Comparando com $T'_{ij} = R_{ik} R_{j\ell} T_{k\ell}$ expandido em $\theta$, concluímos:

$$
[\vec{J}\cdot\hat{n},\, T_{ij}]
=
i\hbar\left(
\varepsilon_{i m k} n_m T_{kj}
+
\varepsilon_{j m k} n_m T_{ik}
\right).
$$

Isso generaliza a relação vetorial:

$$
[J_i, V_j] = i\hbar \varepsilon_{ijk} V_k.
$$

---

## 4.4 Importância física: decomposição em partes irredutíveis

Assim como tensores clássicos podem ser decompostos em partes:

- traço (escalar),
- parte anti-simétrica (vetorial),
- parte simétrica traço-zero (rank 2 “puro”),

os operadores tensoriais também se decompõem em representações **irredutíveis** de SO(3):

$$
1 \otimes 1 = 0 \oplus 1 \oplus 2.
$$

Essa decomposição é crucial porque:
- operadores irredutíveis têm regras simples de rotação,
- levam aos tensores esféricos $T_q^{(k)}$,
- permitem aplicar o Teorema de Wigner–Eckart,
- simplificam cálculos de transições e regras de seleção.

---

## 4.5 Conclusão da seção

> Operadores tensoriais são a generalização natural de escalares e vetores.  
>  
> Um tensor de rank $n$ possui $n$ índices, e **cada índice transforma com sua própria matriz de rotação**.  
>  
> Essa transformação preserva a estrutura geométrica e permite decompor tensores em partes irreducíveis — o que abre caminho para a formulação de tensores esféricos e para toda a teoria de Wigner.

# 4. Operadores Tensoriais

### Intenção conceitual
Tensores aparecem quando combinamos vetores ou derivadas de campos.  
Um tensor de rank 2 tem duas componentes de índice e transforma com dois fatores da matriz de rotação.

---

## 4.1 Definição geral

### Rank 2:

$$
U T_{ij} U^\dagger = \sum_{k\ell} R_{ik} R_{j\ell} \, T_{k\ell}
$$

### Rank arbitrário:

$$
U T_{i_1 i_2 \dots i_n} U^\dagger
= \sum_{j_1\dots j_n}
R_{i_1 j_1} \cdots R_{i_n j_n} T_{j_1\dots j_n}
$$

---

# 5. Decomposição em Componentes Irredutíveis

### Intenção conceitual
O ponto central:  
**tensores cartesianos não são as representações naturais de SO(3)**.  
Eles se decompõem em representações irreducíveis — exatamente como fazemos com produtos tensorais de spins.

Por exemplo:  
$$
1 \otimes 1 = 0 \oplus 1 \oplus 2
$$

---

## 5.1 Exemplo: $ T_{ij} = V_i W_j $

Decompomos:

- **Parte escalar (dim 1)**  
$$
\frac{1}{3} (\vec{V}\cdot\vec{W})\delta_{ij}
$$

- **Parte anti-simétrica (dim 3)** — transforma como vetor  
$$
\frac{V_i W_j - V_j W_i}{2}
$$

- **Parte simétrica traço-zero (dim 5)** — tensor quadrupolar  
$$
\frac{V_i W_j + V_j W_i}{2}
- \frac{1}{3}(\vec{V}\cdot\vec{W})\delta_{ij}
$$

Assim:

$$
\dim 9 = 1 + 3 + 5
$$

Exatamente como esperado da decomposição:

$$
1\otimes 1 = 0 \oplus 1 \oplus 2.
$$

---

# 6. Tensores Esféricos: o truque essencial

### Intenção conceitual
O formalismo cartesiano é horrível para rotações.  
A base natural é a base **esférica**, construída a partir de harmônicos esféricos.

A chave é definir tensores esféricos:

$$
T_q^{(k)} \quad q=-k,\dots,+k
$$

que transformam como:

$$
U T_q^{(k)} U^\dagger
= \sum_{q'} D^{(k)}_{q'q}(\mathcal R)\, T_{q'}^{(k)}.
$$

Esses são os **operadores irreducíveis**.

---

# 7. Conexão com Harmônicos Esféricos

Os harmônicos esféricos satisfazem:

$$
Y_\ell^m(\hat{n}) = \langle \hat{n} | \ell m \rangle
$$

e suas componentes transformam com as matrizes de Wigner $D^{(\ell)}$.

Identificamos operadores vetoriais cartesianos com harmônicos esféricos de $\ell=1$:

$$
T^{(1)}_0 \propto V_z, \qquad
T^{(1)}_{\pm1} \propto V_x \pm i V_y
$$

E tensores quadrupolares com $\ell=2$.

---

# 8. Como um tensor esférico rotaciona?

### Intenção conceitual
Queremos derivar a fórmula fundamental:

$$
U T_q^{(k)} U^\dagger
= \sum_{q'} D^{(k)}_{q'q}(\mathcal R)\, T_{q'}^{(k)}
$$

usando apenas rotações infinitesimais.

---

## 8.1 Começando com uma rotação infinitesimal

$$
U = 1 + \frac{i}{\hbar} \vec{J}\cdot\hat{n}\,\theta
$$

Aplicando:

$$
U T_q^{(k)} U^\dagger
= \left(1+\frac{i}{\hbar} J\cdot\hat{n}\theta\right)
T_q^{(k)}
\left(1-\frac{i}{\hbar} J\cdot\hat{n}\theta\right)
$$

Ficamos com o comutador:

$$
[J\cdot\hat{n}, T_q^{(k)}]
$$

O operador deve transformar dentro do *mesmo* conjunto $\{T_q\}$:

$$
[J\cdot\hat{n}, T_q^{(k)}]
= \sum_{q'} T_{q'}^{(k)} \,
\langle k q' | J\cdot\hat{n} | k q \rangle
$$

Essa é a relação fundamental.

---

# 9. Casos particulares dos comutadores

### 9.1 Rotação em torno de $ \hat{z} $

$$
[J_z, T_q^{(k)}] = \hbar\, q\, T_q^{(k)}
$$

### 9.2 Rotação gerada por $ J_{\pm}=J_x \pm i J_y $

$$
[J_\pm, T_q^{(k)}]
= \hbar\sqrt{(k\mp q)(k\pm q+1)} \; T^{(k)}_{q\pm1}
$$

Essas são as **regras de seleção** dos tensores esféricos.  
Elas imitam perfeitamente as regras de seleção de $|kq\rangle$.

---

# 10. Conclusão

- Vetores cartesianos são tensores esféricos com $k=1$.  
- Quadrupolos são tensores esféricos com $k=2$.  
- Escalares são tensores esféricos com $k=0$.  
- Toda grandeza física pode ser escrita numa base em que sua transformação rotacional é **simples**, dada pela matriz $D^{(k)}$.  
- Esse formalismo é a base de operadores de transição, regras de seleção, Teorema de Wigner–Eckart e álgebra angular completa.
