# 1. Introdução: Por que estudar operadores sob rotações?

O objetivo desta aula é entender **como operadores quânticos se transformam quando aplicamos rotações**.  
Isso é fundamental na mecânica quântica porque:

- estados possuem simetrias,
- Hamiltonianos frequentemente são invariantes rotacionais,
- certos observáveis transformam-se como vetores (posição, momento, spin),
- outros como tensores (quadrupolo elétrico, tensores de spin),
- e alguns permanecem invariantes (energias, integrais escalares).

A ideia é classificar operadores segundo sua **transformação sob o grupo de rotações SO(3)**,
e depois aplicar essa classificação à construção de **tensores esféricos irreducíveis**, que são a base do formalismo de Wigner.

---

# 2. Operadores Escalares

## 2.1 Definição matemática

### Ideia da seção

Aqui queremos ligar duas maneiras de dizer que um operador é **escalar**:

1. **Definição geométrica:** ele é invariável sob qualquer rotação
   $$
   U(\mathcal R) \, K \, U^\dagger(\mathcal R) = K \quad \text{para toda rotação } \mathcal R.
   $$

2. **Definição algébrica:** ele **comuta com todos os componentes** do momento angular
   $$
   [K, J_i] = 0 \quad (i=x,y,z).
   $$

O objetivo é **derivar (2) a partir de (1)**.  
Para isso, olhamos a transformação para uma rotação infinitesimal e usamos uma expansão em série de Taylor do operador de rotação.

---

### Operador de rotação "normal"

Uma rotação finita de ângulo $\boldsymbol{\theta}$ em torno do eixo definido pelo versor $\hat{n}$ é gerada pelo operador de momento angular $\vec{J}$:

$$
U(\mathcal R) = U(\hat{n},\theta)
= \exp\!\left(-\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\, \theta\right).
$$

Aqui:
- $\vec{J}\cdot\hat{n} = J_x n_x + J_y n_y + J_z n_z$ é o componente de $\vec{J}$ na direção de $\hat{n}$;

---

### Por que expandir em série de Taylor?

A condição de escalar
$$
U(\mathcal R) K U^\dagger(\mathcal R) = K
$$
é verdadeira para **qualquer ângulo**.  

Se olharmos especificamente para uma rotação **infinitesimal** (ângulo muito pequeno $\theta$), a transformação fica próxima da identidade e podemos escrever $U$ como:

$$
U = 1 + (\text{algo})\times\theta + O(\theta^2).
$$

O “algo” que multiplica $\theta$ será exatamente o **gerador** da rotação.  
Ao multiplicar $U K U^\dagger$ e manter apenas os termos de primeira ordem em $\theta$, o coeficiente de $\theta$ vira um **comutador** com esse gerador.  
Impor que a transformação não mude $K$ para *todo* $\theta$ nos força a concluir que esse comutador é zero.

---

### Expansão de $U$ até primeira ordem

Expandimos a exponencial:

$$
U(\mathcal R)
= \exp\!\left(-\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
\approx 1 - \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta
+ O(\theta^2).
$$

O adjunto é

$$
U^\dagger(\mathcal R)
= \exp\!\left(+\frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
\approx 1 + \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta
+ O(\theta^2),
$$

pois $\vec{J}$ é hermitiano.

---

### Aplicando a condição de escalar

A condição de invariância é
$$
U K U^\dagger = K.
$$

Substituindo as expansões:

$$
\begin{aligned}
U K U^\dagger
&\approx 
\left(1 - \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right)
K
\left(1 + \frac{i}{\hbar}\,\vec{J}\cdot\hat{n}\,\theta\right) \\
&=
K
- \frac{i}{\hbar}\,(\vec{J}\cdot\hat{n})\,\theta\, K
+ \frac{i}{\hbar}\,K\, (\vec{J}\cdot\hat{n})\,\theta
+ O(\theta^2).
\end{aligned}
$$

Agrupando os termos de primeira ordem em $\theta$:

$$
U K U^\dagger
\approx
K + \frac{i}{\hbar}\,\theta\,
\big(
K\,\vec{J}\cdot\hat{n}
- \vec{J}\cdot\hat{n}\, K
\big)
+ O(\theta^2).
$$

Ou seja,

$$
U K U^\dagger
\approx
K + \frac{i}{\hbar}\,\theta\, [K,\vec{J}\cdot\hat{n}] + O(\theta^2).
$$

Mas, pela definição de operador escalar, devemos ter

$$
U K U^\dagger = K
\quad \text{para qualquer } \theta.
$$

Isso só é possível se o coeficiente de $\theta$ for nulo:

$$
[K,\vec{J}\cdot\hat{n}] = 0.
$$

Como $\hat{n}$ é um versor **arbitrário**, essa igualdade implica que cada componente deve comutar separadamente:

$$
[K,J_x]=[K,J_y]=[K,J_z]=0
\quad \Rightarrow \quad
[K,J_i]=0 \;\; \forall i.
$$

---

### Conclusão desta seção

> **Operadores escalares comutam com todos os componentes do momento angular**.  
> A expansão de $U$ para rotações infinitesimais é a ponte entre:
> - a definição geométrica $U K U^\dagger = K$ e  
> - a condição algébrica de comutação $[K,J_i]=0$.

# 3. Operadores Vetoriais

### Intenção conceitual
Agora introduzimos operadores que **têm direção** (ex: $\vec{r}, \vec{p}, \vec{L}, \vec{S}$).  
Eles devem transformar-se como vetores 3D comuns sob rotações.

---

## 3.1 Transformação vetorial

$$
U(\mathcal R) V_i U^\dagger(\mathcal R) 
= \sum_j R_{ij} \, V_j
$$

Para rotações infinitesimais:

$$
[J_i, V_j] = i\hbar\, \varepsilon_{ijk} V_k
$$

que é a estrutura do grupo SO(3).

---

# 4. Operadores Tensoriais

### Intenção conceitual
Tensores aparecem quando combinamos vetores ou derivadas de campos.  
Um tensor de rank 2 tem duas componentes de índice e transforma com dois fatores da matriz de rotação.

---

## 4.1 Definição geral

### Rank 2:

$$
U T_{ij} U^\dagger = \sum_{k\ell} R_{ik} R_{j\ell} \, T_{k\ell}
$$

### Rank arbitrário:

$$
U T_{i_1 i_2 \dots i_n} U^\dagger
= \sum_{j_1\dots j_n}
R_{i_1 j_1} \cdots R_{i_n j_n} T_{j_1\dots j_n}
$$

---

# 5. Decomposição em Componentes Irredutíveis

### Intenção conceitual
O ponto central:  
**tensores cartesianos não são as representações naturais de SO(3)**.  
Eles se decompõem em representações irreducíveis — exatamente como fazemos com produtos tensorais de spins.

Por exemplo:  
$$
1 \otimes 1 = 0 \oplus 1 \oplus 2
$$

---

## 5.1 Exemplo: $ T_{ij} = V_i W_j $

Decompomos:

- **Parte escalar (dim 1)**  
$$
\frac{1}{3} (\vec{V}\cdot\vec{W})\delta_{ij}
$$

- **Parte anti-simétrica (dim 3)** — transforma como vetor  
$$
\frac{V_i W_j - V_j W_i}{2}
$$

- **Parte simétrica traço-zero (dim 5)** — tensor quadrupolar  
$$
\frac{V_i W_j + V_j W_i}{2}
- \frac{1}{3}(\vec{V}\cdot\vec{W})\delta_{ij}
$$

Assim:

$$
\dim 9 = 1 + 3 + 5
$$

Exatamente como esperado da decomposição:

$$
1\otimes 1 = 0 \oplus 1 \oplus 2.
$$

---

# 6. Tensores Esféricos: o truque essencial

### Intenção conceitual
O formalismo cartesiano é horrível para rotações.  
A base natural é a base **esférica**, construída a partir de harmônicos esféricos.

A chave é definir tensores esféricos:

$$
T_q^{(k)} \quad q=-k,\dots,+k
$$

que transformam como:

$$
U T_q^{(k)} U^\dagger
= \sum_{q'} D^{(k)}_{q'q}(\mathcal R)\, T_{q'}^{(k)}.
$$

Esses são os **operadores irreducíveis**.

---

# 7. Conexão com Harmônicos Esféricos

Os harmônicos esféricos satisfazem:

$$
Y_\ell^m(\hat{n}) = \langle \hat{n} | \ell m \rangle
$$

e suas componentes transformam com as matrizes de Wigner $D^{(\ell)}$.

Identificamos operadores vetoriais cartesianos com harmônicos esféricos de $\ell=1$:

$$
T^{(1)}_0 \propto V_z, \qquad
T^{(1)}_{\pm1} \propto V_x \pm i V_y
$$

E tensores quadrupolares com $\ell=2$.

---

# 8. Como um tensor esférico rotaciona?

### Intenção conceitual
Queremos derivar a fórmula fundamental:

$$
U T_q^{(k)} U^\dagger
= \sum_{q'} D^{(k)}_{q'q}(\mathcal R)\, T_{q'}^{(k)}
$$

usando apenas rotações infinitesimais.

---

## 8.1 Começando com uma rotação infinitesimal

$$
U = 1 + \frac{i}{\hbar} \vec{J}\cdot\hat{n}\,\theta
$$

Aplicando:

$$
U T_q^{(k)} U^\dagger
= \left(1+\frac{i}{\hbar} J\cdot\hat{n}\theta\right)
T_q^{(k)}
\left(1-\frac{i}{\hbar} J\cdot\hat{n}\theta\right)
$$

Ficamos com o comutador:

$$
[J\cdot\hat{n}, T_q^{(k)}]
$$

O operador deve transformar dentro do *mesmo* conjunto $\{T_q\}$:

$$
[J\cdot\hat{n}, T_q^{(k)}]
= \sum_{q'} T_{q'}^{(k)} \,
\langle k q' | J\cdot\hat{n} | k q \rangle
$$

Essa é a relação fundamental.

---

# 9. Casos particulares dos comutadores

### 9.1 Rotação em torno de $ \hat{z} $

$$
[J_z, T_q^{(k)}] = \hbar\, q\, T_q^{(k)}
$$

### 9.2 Rotação gerada por $ J_{\pm}=J_x \pm i J_y $

$$
[J_\pm, T_q^{(k)}]
= \hbar\sqrt{(k\mp q)(k\pm q+1)} \; T^{(k)}_{q\pm1}
$$

Essas são as **regras de seleção** dos tensores esféricos.  
Elas imitam perfeitamente as regras de seleção de $|kq\rangle$.

---

# 10. Conclusão

- Vetores cartesianos são tensores esféricos com $k=1$.  
- Quadrupolos são tensores esféricos com $k=2$.  
- Escalares são tensores esféricos com $k=0$.  
- Toda grandeza física pode ser escrita numa base em que sua transformação rotacional é **simples**, dada pela matriz $D^{(k)}$.  
- Esse formalismo é a base de operadores de transição, regras de seleção, Teorema de Wigner–Eckart e álgebra angular completa.
