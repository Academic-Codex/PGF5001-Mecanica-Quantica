# Aula 21 — Estrutura Fina do Átomo de Hidrogênio

Nesta aula estudamos *correções relativísticas* ao Hamiltoniano do átomo de hidrogênio.  
O objetivo conceitual é entender:

1. De onde vêm as três correções responsáveis pela estrutura fina:  
   - energia cinética relativística (RKE)  
   - acoplamento spin–órbita (SO)  
   - termo de Darwin (D)  

2. Por que essas correções surgem naturalmente da expansão relativística da energia.

3. Como a teoria de perturbação degenerada entra no problema — já que o Hamiltoniano não relativístico do hidrogênio é altamente degenerado.

A ideia final é obter a expressão energética:

$$
E_{n j} = -\frac{Z}{2 n^2}
\left[
1 + (Z\alpha)^2 \frac{1}{n^2}
\left( \frac34 - \frac{n}{j + 1/2} \right)
+ \cdots
\right]
$$

que explica o desdobramento fino observado experimentalmente.

## 1. Hamiltoniano de referência

O Hamiltoniano do hidrogênio (não relativístico) é:

$$
H_0 = \frac{\hat{p}^2}{2m} + V(r)
= \frac{\hat{p}^2}{2m} - \frac{Z e^2}{r}.
$$

Os autoestados são os conhecidos $|n\ell m_\ell\rangle$.

A energia depende **somente de $n$**:

$$
E_n = -\frac{m c^2 Z^2 \alpha^2}{2 n^2}.
$$

**Intenção pedagógica:**  
O professor destaca a simetria “acidental” do hidrogênio: a energia não depende de $\ell$ nem de $m_\ell$.  
Essa degenerescência é o motivo pelo qual precisaremos usar **perturbação degenerada** para a estrutura fina.

## 2. Correções de Estrutura Fina

As correções de ordem $\alpha^4$ vêm de três termos:

1. **Energia cinética relativística**  
   $H_{\mathrm{RKE}}$

2. **Acoplamento spin–órbita**  
   $H_{\mathrm{SO}}$

3. **Termo de Darwin** (efeito puramente relativístico associado ao Zitterbewegung)  
   $H_D$

Assim:

$$
H_{\mathrm{FS}} = H_{\mathrm{RKE}} + H_{\mathrm{SO}} + H_D.
$$

Todas as três correções são da ordem de $(Z\alpha)^4$.

## 3. Correção relativística da energia cinética

Relatividade especial:

$$
E = \sqrt{m^2 c^4 + c^2 p^2}.
$$

Expandindo para $p \ll mc$:

$$
E \approx m c^2 + \frac{p^2}{2m} - \frac{p^4}{8 m^3 c^2} + \cdots
$$

Comparando com o termo não-relativístico, a correção é:

$$
H_{\mathrm{RKE}} = -\frac{p^4}{8 m^3 c^2}.
$$

**Intenção pedagógica:**  
A expansão mostra que a Cinética NR ignora termos de ordem superior em $p^2$. A correção negativamente assinada diz que o elétron relativístico é *menos ligado* do que o previsto pelo modelo NR.

## 4. Acoplamento Spin–Órbita

Clássica: o elétron em movimento vê o núcleo se movendo e, portanto, observa um campo magnético efetivo.

O Hamiltoniano:

$$
H_{\mathrm{SO}} = -\frac{1}{2} \boldsymbol{\mu}\cdot \mathbf{B}'.
$$

Onde:

- $\boldsymbol{\mu}$ = momento magnético do elétron  
- $\mathbf{B}'$ = campo magnético no referencial do elétron (efeito de Lorentz)

Usando:

$$
\mathbf{E} = -\nabla V(r),
\qquad
\mathbf{B}' = \frac{1}{c}\mathbf{E} \times \mathbf{v}
= \frac{1}{2m^2 c^2 r} \frac{dV}{dr}\, \mathbf{L}
$$

obtém-se o operador:

$$
H_{\mathrm{SO}} =
\frac{1}{2m^2 c^2 r}\frac{dV}{dr}\, \mathbf{L}\cdot\mathbf{S}.
$$

Para o potencial Coulombiano:

$$
\frac{1}{r}\frac{dV}{dr} = \frac{Z e^2}{r^3}.
$$

Então:

$$
H_{\mathrm{SO}} = 
\frac{Z e^2}{2 m^2 c^2 r^3}\, \mathbf{L}\cdot\mathbf{S}.
$$

**Ideia conceitual:**  
O acoplamento spin–órbita mistura não apenas $\ell$ e $m_\ell$, mas também o spin.  
Por isso o professor enfatiza que devemos trabalhar no espaço:

$$
\mathcal{H} = \mathcal{H}_{\text{orb}} \otimes \mathcal{H}_{\text{spin}}
$$

e usar autoestados de **momento angular total** $|n\ell j m_j\rangle$.

## 5. Termo de Darwin (Zitterbewegung)

O termo relativístico adicional:

$$
H_D = \frac{\hbar^2}{8 m^2 c^2}\nabla^2 V.
$$

Para o potencial Coulombiano:

$$
V(r) = -\frac{Z e^2}{r},
\qquad
\nabla^2 V = 4\pi Z e^2 \delta(\mathbf{r}).
$$

Então:

$$
H_D =
\frac{\pi Z e^2 \hbar^2}{2 m^2 c^2}\delta(\mathbf{r}).
$$

**Interpretação física:**  
Aparece porque na equação de Dirac há *tremores* quânticos do elétron (Zitterbewegung).  
O termo só contribui para **s-estados** ($\ell=0$), pois apenas eles têm $\psi(0)\neq 0$.

No hidrogênio:

$$
\langle H_D\rangle
= Z\alpha^2 (-E_n)\frac{1}{n} \delta_{\ell 0}.

## 6. Por que usar perturbação degenerada?

O Hamiltoniano $H_0$ é degenerado em relação a $\ell$ e $m_\ell$.

O subespaço degenerado é:

$$
|n\ell m_\ell m_s\rangle
= |n\ell m_\ell\rangle \otimes |s m_s\rangle.
$$

Para diagonalizar $H_{\mathrm{FS}}$:

- verificamos com quais operadores cada termo comuta  
- apenas $H_{\mathrm{SO}}$ não comuta com $L_z$ nem com $S_z$  
- mas **todos** comutam com $J^2$ e $J_z$  

Portanto, o professor muda a base para:

$$
|n\ell j m_j\rangle.
$$

que é a base natural para diagonalizar $H_{\mathrm{SO}}$.

## 7. Aplicação do Teorema de Wigner–Eckart

Para o termo relativístico:

$$
H_{\mathrm{RKE}} = -\frac{\alpha^2}{8}p^4
$$

ele é **independente de spin**, logo:

$$
\langle n\ell m_\ell m_s|H_{\mathrm{RKE}}|n\ell m_\ell' m_s'\rangle
= \delta_{m_s m_s'} \langle n\ell m_\ell|H_{\mathrm{RKE}}|n\ell m_\ell'\rangle.
$$

Usando Wigner–Eckart:

$$
\langle n \ell m_\ell|H_{\mathrm{RKE}}|n\ell m_\ell'\rangle
= \delta_{m_\ell m_\ell'} \langle n\ell 0|H_{\mathrm{RKE}}|n\ell 0\rangle.
$$

Quando mudamos para $|n\ell j m_j\rangle$, a soma dos coeficientes de CG fornece unidade, resultando em:

$$
\langle n\ell j m_j|H_{\mathrm{RKE}}|n\ell j m_j\rangle
= \langle n \ell 0|H_{\mathrm{RKE}}|n\ell 0\rangle.

## 8. Valores esperados necessários

Em unidades atômicas ($m=e=\hbar=1$):

$$
\left\langle \frac{1}{r}\right\rangle = \frac{Z}{n^2},
\qquad
\left\langle \frac{1}{r^2}\right\rangle = \frac{Z^2}{n^2(\ell + 1/2)}.
$$

$$
\left\langle \frac{1}{r^3}\right\rangle
= \frac{Z^3}{n^3(\ell + 1/2)\ell(\ell+1)}.
$$

Mas esse último **diverge para $\ell=0$**, reforçando que o termo SO não contribui para estados s.

## 9. Resultados separados

### (1) Energia cinética relativística
$$
\langle H_{\mathrm{RKE}} \rangle
= (Z\alpha)^2(-E_n)\frac{1}{n^2}
\left(\frac34 - \frac{n}{\ell + 1/2}\right)
$$

### (2) Termo de Darwin
Aparece apenas para $\ell=0$:

$$
\langle H_D \rangle
= (Z\alpha)^2 (-E_n)\frac{1}{n}\delta_{\ell 0}.
$$

### (3) Acoplamento spin–órbita

Usando:

$$
\mathbf{L}\cdot \mathbf{S}
= \frac{1}{2}(j(j+1) - \ell(\ell+1) - s(s+1)),
$$

temos:

$$
\langle H_{\mathrm{SO}}\rangle
= (Z\alpha)^2(-E_n)\frac{1}{2n^3}
\left[
\frac{j(j+1)-\ell(\ell+1)-3/4}{\ell(\ell+1/2)(\ell+1)}
\right].

## 10. Correção completa da estrutura fina

Somando os três termos:

$$
\Delta E_{\mathrm{FS}}
=
(Z\alpha)^2(-E_n)\frac{1}{n^2}
\left(
\frac34 - \frac{n}{j + 1/2}
\right).
$$

Portanto:

$$
E_{n j}
=
-\frac{Z}{2n^2}
\left[
1
+
(Z\alpha)^2 \frac{1}{n^2}
\left(
\frac34 - \frac{n}{j+1/2}
\right)
+ \cdots
\right].
$$

**Moral conceitual:**  
A degenerescência em $\ell$ do hidrogênio é quebrada pela estrutura fina, mas a simetria ainda preserva a dependência apenas de $j$.

A energia passa a depender de $j$ — e isso bate perfeitamente com o espectro experimental.
