# Métodos de Aproximação — Teoria de Perturbações Não-Degenerada

---

## 1. Motivação

Em muitos sistemas quânticos, o Hamiltoniano total $ H $ não pode ser resolvido exatamente.  
Nesses casos, recorremos a **métodos aproximativos**, que se dividem em:

1. **Métodos numéricos**
2. **Teoria de Perturbações**, aplicável quando $ H $ difere pouco de um Hamiltoniano conhecido $ H_0 $.

---

## 2. Estrutura do Problema

Escrevemos o Hamiltoniano como:

$$
H = H_0 + \lambda H_1, \quad \lambda \ll 1
$$

onde:

- $ H_0 $: Hamiltoniano não perturbado, com soluções conhecidas  
  $ H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle $
- $ H_1 $: pequena perturbação  
- $ \lambda $: parâmetro de controle da expansão

Buscamos soluções em **séries de potência em $ \lambda $**:

$$
\begin{aligned}
E_n &= E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots \\
|n\rangle &= |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \dots
\end{aligned}
$$

---


## 3. Inserindo na Equação de Schrödinger

$$
H |n\rangle = E_n |n\rangle
$$

Substituindo $ H = H_0 + \lambda H_1 $ e expandindo:

$$
(H_0 + \lambda H_1)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \dots)
= (E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \dots)
$$

Igualamos **ordem a ordem em $ \lambda $**.

---


## 4. Ordem Zero $ (\lambda^0) $

$$
H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle
$$

✅ Equação do sistema não perturbado (solução conhecida).

---


## 5. Primeira Ordem $ (\lambda^1) $

$$
H_0 |n^{(1)}\rangle + H_1 |n^{(0)}\rangle = E_n^{(0)} |n^{(1)}\rangle + E_n^{(1)} |n^{(0)}\rangle
$$

Reorganizando:

$$
(H_0 - E_n^{(0)})|n^{(1)}\rangle = (E_n^{(1)} - H_1)|n^{(0)}\rangle
$$

---



## 6. Primeiro Truque: Projeção Ortogonal

Multiplicando à esquerda por $ \langle n^{(0)}| $ e usando ortonormalidade:

$$
\langle n^{(0)}|H_0|n^{(1)}\rangle + \langle n^{(0)}|H_1|n^{(0)}\rangle =
E_n^{(0)}\langle n^{(0)}|n^{(1)}\rangle + E_n^{(1)}\langle n^{(0)}|n^{(0)}\rangle
$$

Escolhemos a **normalização**:

$$
\langle n^{(0)}|n\rangle = 1 \quad \Rightarrow \quad \langle n^{(0)}|n^{(1)}\rangle = 0
$$

Assim:

$$
E_n^{(1)} = \langle n^{(0)}|H_1|n^{(0)}\rangle
$$

✅ A primeira correção à energia é o valor médio da perturbação no estado base.

---


## 7. Segundo Truque: Operador Projetor $ \hat{\phi}_n $

O problema com a inversão de $ (E_n^{(0)} - H_0) $ é que ele **não é invertível** sobre o subespaço gerado por $ |n^{(0)}\rangle $.  
Para contornar isso, definimos o **projetor ortogonal**:

$$
\hat{\phi}_n = \hat{1} - |n^{(0)}\rangle \langle n^{(0)}|
$$

Esse operador **remove a componente paralela** a $ |n^{(0)}\rangle $, permitindo aplicar o inverso de $ E_n^{(0)} - H_0 $ de forma bem definida.

Aplicando o projetor:

$$
|n^{(1)}\rangle = \hat{\phi}_n \frac{1}{E_n^{(0)} - H_0} (H_1 - E_n^{(1)}) |n^{(0)}\rangle
$$

Como o termo $ E_n^{(1)}|n^{(0)}\rangle $ é eliminado pela projeção, temos:

$$
|n^{(1)}\rangle = \sum_{k\ne n} |k^{(0)}\rangle 
\frac{\langle k^{(0)}|H_1|n^{(0)}\rangle}{E_n^{(0)} - E_k^{(0)}}
$$

---

## 8. Segunda Ordem

A ordem $ \lambda^2 $ fornece:

$$
(H_0 - E_n^{(0)})|n^{(2)}\rangle = (E_n^{(1)} - H_1)|n^{(1)}\rangle + E_n^{(2)}|n^{(0)}\rangle
$$

Multiplicando à esquerda por $ \langle n^{(0)}| $, obtemos:

$$
E_n^{(2)} = \sum_{k \neq n} 
\frac{|\langle k^{(0)}|H_1|n^{(0)}\rangle|^2}{E_n^{(0)} - E_k^{(0)}}
$$

---

## 9. Resultados Finais

$$
\boxed{
\begin{aligned}
E_n &= E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots \\
E_n^{(1)} &= \langle n^{(0)}|H_1|n^{(0)}\rangle \\
E_n^{(2)} &= \sum_{k\neq n} \frac{|\langle k^{(0)}|H_1|n^{(0)}\rangle|^2}{E_n^{(0)} - E_k^{(0)}} \\
|n^{(1)}\rangle &= \sum_{k\neq n} |k^{(0)}\rangle \frac{\langle k^{(0)}|H_1|n^{(0)}\rangle}{E_n^{(0)} - E_k^{(0)}}
\end{aligned}
}
$$

---


## 10. Truques Conceituais e Matemáticos ✨

| Truque | Propósito |
|:--|:--|
| **Expansão em série de Taylor (ou potência)** | Tratar $ H_1 $ como pequena e extrair correções ordem a ordem |
| **Projeção ortogonal $ \hat{\phi}_n $** | Elimina a componente do espaço onde $ H_0 $- $E_n^{(0)} $ não é invertível |
| **Normalização $ \langle n^{(0)} \|n\rangle = 1 $** | Garante consistência e simplifica os cálculos |
| **Separação ordem a ordem em $ \lambda $** | Isola contribuições progressivas das correções |
| **Uso da resolução da identidade** | Permite reescrever o inverso de operadores como somas explícitas |
| **Analogia com séries convergentes e assintóticas** | Mostra quando a expansão perturbativa é válida ou falha |

---


## 11. Observações sobre Convergência

- Nem toda série perturbativa **converge** — mas pode ser **assintoticamente útil**.  
- Exemplo: aproximação de Stirling  

  $$
  \Gamma(x) \approx \sqrt{2\pi}x^{x - \frac{1}{2}} e^{-x}\left(1 + \frac{1}{12x} + \frac{1}{288x^2} + \dots\right)
  $$

  Embora divergente, fornece excelente aproximação até certa ordem.

- Um sistema é dito **não-perturbativo** quando o desenvolvimento em $ \lambda $ falha, como em:

  $$
  e^{-1/x^2} = 0 + 0x + 0x^2 + 0x^3 + \dots
  $$

  Nenhuma expansão perturbativa capta o comportamento — apenas abordagens não-perturbativas (ex: instantons, integrais de caminho).

---

**→ Conclusão:**  
A Teoria de Perturbações permite obter correções sistemáticas às energias e estados de um sistema a partir de uma solução conhecida.  
O uso de projeções, normalizações e expansões ordenadas em $ \lambda $ é o coração do método — e a análise de convergência é o que define o domínio de validade física da aproximação.
