# Métodos de Aproximação — Teoria de Perturbações Não-Degenerada

---

## 1. Motivação

Em muitos sistemas quânticos, o Hamiltoniano total $ H $ não pode ser resolvido exatamente.  
Nesses casos, recorremos a **métodos aproximativos**, que se dividem em:

1. **Métodos numéricos**
2. **Teoria de Perturbações**, aplicável quando $ H $ difere pouco de um Hamiltoniano conhecido $ H_0 $.

---

## 2. Estrutura do Problema

Escrevemos o Hamiltoniano como:

$$
H = H_0 + \lambda H_1, \quad \lambda \ll 1
$$

onde:

- $ H_0 $: Hamiltoniano não perturbado, com soluções conhecidas  
  $ H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle $
- $ H_1 $: pequena perturbação  
- $ \lambda $: parâmetro de controle da expansão

Buscamos soluções em **séries de potência em $ \lambda $**:

$$
\begin{aligned}
E_n &= E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots \\
|n\rangle &= |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \dots
\end{aligned}
$$

---


## 3. Inserindo na Equação de Schrödinger

$$
H |n\rangle = E_n |n\rangle
$$

Substituindo $ H = H_0 + \lambda H_1 $ e expandindo:

$$
(H_0 + \lambda H_1)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \dots)
= (E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \dots)
$$

Igualamos **ordem a ordem em $ \lambda $**.

---


## 4. Ordem Zero $ (\lambda^0) $

$$
H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle
$$

✅ Equação do sistema não perturbado (solução conhecida).

---


## 5. Primeira Ordem $ (\lambda^1) $

$$
H_0 |n^{(1)}\rangle + H_1 |n^{(0)}\rangle = E_n^{(0)} |n^{(1)}\rangle + E_n^{(1)} |n^{(0)}\rangle
$$

Reorganizando:

$$
(H_0 - E_n^{(0)})|n^{(1)}\rangle = (E_n^{(1)} - H_1)|n^{(0)}\rangle
$$

---



## 6. Primeiro Truque: Projeção Ortogonal

Multiplicando à esquerda por $ \langle n^{(0)}| $ e usando ortonormalidade:

$$
\langle n^{(0)}|H_0|n^{(1)}\rangle + \langle n^{(0)}|H_1|n^{(0)}\rangle =
E_n^{(0)}\langle n^{(0)}|n^{(1)}\rangle + E_n^{(1)}\langle n^{(0)}|n^{(0)}\rangle
$$

Escolhemos a **normalização**:

$$
\langle n^{(0)}|n\rangle = 1 \quad \Rightarrow \quad \langle n^{(0)}|n^{(1)}\rangle = 0
$$

Assim:

$$
E_n^{(1)} = \langle n^{(0)}|H_1|n^{(0)}\rangle
$$

✅ A primeira correção à energia é o valor médio da perturbação no estado base.

---


## 7. Segundo Truque: Operador Projetor $ \hat{\phi}_n $

O problema com a inversão de $ (E_n^{(0)} - H_0) $ é que ele **não é invertível** sobre o subespaço gerado por $ |n^{(0)}\rangle $.  
Para contornar isso, definimos o **projetor ortogonal**:

$$
\hat{\phi}_n = \hat{1} - |n^{(0)}\rangle \langle n^{(0)}|
$$

Esse operador **remove a componente paralela** a $ |n^{(0)}\rangle $, permitindo aplicar o inverso de $ E_n^{(0)} - H_0 $ de forma bem definida.

Aplicando o projetor:

$$
|n^{(1)}\rangle = \hat{\phi}_n \frac{1}{E_n^{(0)} - H_0} (H_1 - E_n^{(1)}) |n^{(0)}\rangle
$$

Como o termo $ E_n^{(1)}|n^{(0)}\rangle $ é eliminado pela projeção, temos:

$$
|n^{(1)}\rangle = \sum_{k\ne n} |k^{(0)}\rangle 
\frac{\langle k^{(0)}|H_1|n^{(0)}\rangle}{E_n^{(0)} - E_k^{(0)}}
$$

---

## 8. Segunda Ordem

A ordem $ \lambda^2 $ fornece:

$$
(H_0 - E_n^{(0)})|n^{(2)}\rangle = (E_n^{(1)} - H_1)|n^{(1)}\rangle + E_n^{(2)}|n^{(0)}\rangle
$$

Multiplicando à esquerda por $ \langle n^{(0)}| $, obtemos:

$$
E_n^{(2)} = \sum_{k \neq n} 
\frac{|\langle k^{(0)}|H_1|n^{(0)}\rangle|^2}{E_n^{(0)} - E_k^{(0)}}
$$

---

## 9. Resultados Finais

$$
\boxed{
\begin{aligned}
E_n &= E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \dots \\
E_n^{(1)} &= \langle n^{(0)}|H_1|n^{(0)}\rangle \\
E_n^{(2)} &= \sum_{k\neq n} \frac{|\langle k^{(0)}|H_1|n^{(0)}\rangle|^2}{E_n^{(0)} - E_k^{(0)}} \\
|n^{(1)}\rangle &= \sum_{k\neq n} |k^{(0)}\rangle \frac{\langle k^{(0)}|H_1|n^{(0)}\rangle}{E_n^{(0)} - E_k^{(0)}}
\end{aligned}
}
$$

---


## 10. Truques Conceituais e Matemáticos ✨

| Truque | Propósito |
|:--|:--|
| **Expansão em série de Taylor (ou potência)** | Tratar $ H_1 $ como pequena e extrair correções ordem a ordem |
| **Projeção ortogonal $ \hat{\phi}_n $** | Elimina a componente do espaço onde $ H_0 $- $E_n^{(0)} $ não é invertível |
| **Normalização $ \langle n^{(0)} \|n\rangle = 1 $** | Garante consistência e simplifica os cálculos |
| **Separação ordem a ordem em $ \lambda $** | Isola contribuições progressivas das correções |
| **Uso da resolução da identidade** | Permite reescrever o inverso de operadores como somas explícitas |
| **Analogia com séries convergentes e assintóticas** | Mostra quando a expansão perturbativa é válida ou falha |

---


## 11. Observações sobre Convergência

- Nem toda série perturbativa **converge** — mas pode ser **assintoticamente útil**.  
- Exemplo: aproximação de Stirling  

  $$
  \Gamma(x) \approx \sqrt{2\pi}x^{x - \frac{1}{2}} e^{-x}\left(1 + \frac{1}{12x} + \frac{1}{288x^2} + \dots\right)
  $$

  Embora divergente, fornece excelente aproximação até certa ordem.

- Um sistema é dito **não-perturbativo** quando o desenvolvimento em $ \lambda $ falha, como em:

  $$
  e^{-1/x^2} = 0 + 0x + 0x^2 + 0x^3 + \dots
  $$

  Nenhuma expansão perturbativa capta o comportamento — apenas abordagens não-perturbativas (ex: instantons, integrais de caminho).

---

**→ Conclusão:**  
A Teoria de Perturbações permite obter correções sistemáticas às energias e estados de um sistema a partir de uma solução conhecida.  
O uso de projeções, normalizações e expansões ordenadas em $ \lambda $ é o coração do método — e a análise de convergência é o que define o domínio de validade física da aproximação.

# Método Variacional

---

## 1. Motivação

O método variacional é uma técnica geral para **estimar a energia do estado fundamental** de um sistema quântico, mesmo quando o Hamiltoniano **não pode ser resolvido exatamente**.

Diferente da teoria de perturbações, **não requeremos conhecer** uma solução exata de $ H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle $.

Ele é aplicável inclusive a **sistemas não-perturbativos**, e fornece **um limite superior** para a energia fundamental $ E_0 $.

---

## 2. Formulação Geral

Seja um Hamiltoniano qualquer $ H $ e um estado arbitrário, normalizado ou não, $ |\tilde{0}\rangle $.

Definimos o **valor esperado (energia média):**

$$
\tilde{E} = \frac{\langle \tilde{0}|H|\tilde{0}\rangle}{\langle \tilde{0}|\tilde{0}\rangle}
$$

O **Princípio Variacional** afirma que:

$$
\tilde{E} \ge E_0
$$

onde $ E_0 $ é a energia do estado fundamental.

---

## 3. Demonstração

Assumindo que o conjunto de autovetores $ \{|k\rangle\} $ de $ H $ forma uma base completa:

$$
H|k\rangle = E_k |k\rangle, \quad k=0,1,2,\dots
$$

Expandimos o estado arbitrário:

$$
|\tilde{0}\rangle = \sum_{k=0}^\infty |k\rangle \langle k|\tilde{0}\rangle
$$

Substituindo na expressão de $ \tilde{E} $:

$$
\tilde{E} 
= \frac{\sum_k |\langle k|\tilde{0}\rangle|^2 E_k}
        {\sum_k |\langle k|\tilde{0}\rangle|^2}
$$

Como $ E_k \ge E_0 $ para todo $ k $, podemos escrever:

$$
\tilde{E} 
= E_0 + 
  \frac{\sum_k |\langle k|\tilde{0}\rangle|^2 (E_k - E_0)}
       {\sum_k |\langle k|\tilde{0}\rangle|^2}
$$

O termo adicional é **positivo ou nulo**, logo:

$$
\tilde{E} \ge E_0
$$

---

## 4. Implementação Prática

Escolhe-se uma **função de teste variacional** $ |\tilde{0}(\lambda_1, \lambda_2, \dots)\rangle $  
dependente de **parâmetros variacionais** $ \{\lambda_i\} $.

A energia média correspondente é:

$$
\tilde{E}(\lambda_1, \lambda_2, \dots) =
\frac{\langle \tilde{0}(\lambda)|H|\tilde{0}(\lambda)\rangle}
     {\langle \tilde{0}(\lambda)|\tilde{0}(\lambda)\rangle}
$$

Para obter a melhor estimativa de $ E_0 $, **minimizamos** essa função:

$$
\frac{\partial \tilde{E}}{\partial \lambda_1} = 
\frac{\partial \tilde{E}}{\partial \lambda_2} = 
\dots = 0
$$

O conjunto de parâmetros $ \{\lambda_i^\ast\} $ que satisfaz essas condições dá a melhor aproximação do estado fundamental.

---

## 5. Estados Excitados

Para estimar **estados excitados**, escolhemos funções de teste **ortogonais** ao estado fundamental encontrado:

$$
\langle \tilde{0}(\lambda^\ast) | \tilde{1}(\mu)\rangle = 0
$$

Em seguida, minimizamos novamente a energia média:

$$
\frac{\partial \tilde{E}_1}{\partial \mu_i} = 0
$$

---

## 6. Resumo Conceitual 

| Conceito | Significado |
|:--|:--|
| **Estado de teste** $ |\tilde{0}\rangle $ | Aproximação para o estado real do sistema |
| **Energia variacional** $ \tilde{E} $ | Valor esperado da energia; fornece limite superior de $ E_0 $ |
| **Minimização** | Garante a melhor aproximação possível dentro da família escolhida |
| **Ortogonalização** | Necessária para acessar estados excitados |
| **Vantagem** | Não requer decomposição perturbativa; útil em sistemas fortemente interagentes |

---

**Conclusão:**  
O método variacional fornece uma forma poderosa e intuitiva de estimar a energia fundamental e, com restrições adicionais, também energias excitadas.  
A escolha adequada da função de teste e dos parâmetros é crucial para obter aproximações confiáveis — esse é o mesmo princípio explorado em abordagens modernas como **VQE (Variational Quantum Eigensolver)** em computação quântica.

# Teoria de Perturbação Independente do Tempo (caso não degenerado)

## 1. Setup do problema

Considere um Hamiltoniano que pode ser escrito como

$$
H(\lambda) = H_0 + \lambda H_1,
$$

onde:

- $H_0$ é um Hamiltoniano que sabemos resolver exatamente;
- $H_1$ é uma perturbação "pequena";
- $\lambda$ é um parâmetro adimensional que controla a força da perturbação ($\lambda \ll 1$).

Suponha que o espectro de $H_0$ seja **não degenerado**:

$$
H_0 \, |n^{(0)}\rangle = E_n^{(0)} \, |n^{(0)}\rangle,
\qquad
E_n^{(0)} \neq E_m^{(0)} \text{ se } n \neq m.
$$

Queremos os autovalores e autovetores de $H(\lambda)$, isto é, resolver

$$
H(\lambda)\,|n(\lambda)\rangle = E_n(\lambda)\,|n(\lambda)\rangle.
$$

---



## 2. Ansatz de séries em potências assintótica em $\lambda$

A hipótese central da teoria de perturbação é que os autovalores e autovetores de $H(\lambda)$ podem ser expandidos em séries assintótica de potências de $\lambda$:

$$
E_n(\lambda) = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots,
$$

$$
|n(\lambda)\rangle = |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots.
$$

Aqui:

- $E_n^{(0)}$, $|n^{(0)}\rangle$ são o autovalor e autovetor do problema não perturbado;
- $E_n^{(1)}, E_n^{(2)}, \dots$ são as correções de 1ª, 2ª, ... ordem;
- $|n^{(1)}\rangle, |n^{(2)}\rangle,\dots$ são as correções ao estado.

---

## Séries Assintóticas em Mecânica Quântica

Uma **série de perturbação** tem a forma formal:

$$
E(λ) ∼ Σ_{k=0}^{∞} E^{(k)} λ^{k}   quando λ → 0.
$$
Essa expressão **parece** uma expansão de Taylor, mas é, na verdade,  
uma **série assintótica**. Suas principais propriedades são:

- ❌ **Não tem convergência garantida.**  
  Mesmo para λ diferente de zero, a série pode divergir totalmente.

- ❌ **Não é obtida por derivadas da função real E(λ).**  
  Os coeficientes $E^{(k)}$ não são $f^{(k)}(0)/k!$, mas aparecem ao resolver  
  equações de projeção no espaço de Hilbert.

- ❌ **Não permite reconstruir a solução exata para $\lambda$ finito.**  
  Em geral, somar ordens muito altas piora a aproximação.

- ✔ **É válida apenas como aproximação no limite λ → 0.**  
  Quanto menor $\lambda$, mais precisa é a aproximação com poucas ordens.

- ✔ **Os primeiros termos dão a melhor previsão física.**  
  Em séries assintóticas, existe uma ordem “ótima” para truncar.

- ✔ **É a ferramenta prática fundamental em MQ, QED, moléculas, etc.**  
  Mesmo divergentes, essas séries fornecem previsões com alta precisão.


## 3. Inserindo o ansatz na equação de Schrödinger

Começamos de

$$
H(\lambda)\,|n(\lambda)\rangle = E_n(\lambda)\,|n(\lambda)\rangle.
$$

Substituindo $H(\lambda) = H_0 + \lambda H_1$ e as expansões para $E_n(\lambda)$ e $|n(\lambda)\rangle$:

$$
\bigl(H_0 + \lambda H_1\bigr)
\left( |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots \right)
=
\left(E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots\right)
\left( |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots \right).
$$

Agora expandimos ambos os lados em potências de $\lambda$ e igualamos ordem a ordem.

---


### Desenvolvimento da teoria de perturbação (caso não degenerado)

Começamos com o Hamiltoniano perturbado, o estado e a energia escritos como séries em $\lambda$:

$$
H(\lambda) = H_0 + \lambda H_1,
$$

$$
|n(\lambda)\rangle = |n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots,
$$

$$
E_n(\lambda) = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots.
$$

Substituímos tudo na equação de autovalor

$$
H(\lambda)|n(\lambda)\rangle = E_n(\lambda)|n(\lambda)\rangle,
$$

isto é,

$$
(H_0 + \lambda H_1)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots)
= (E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots)(|n^{(0)}\rangle + \lambda |n^{(1)}\rangle + \lambda^2 |n^{(2)}\rangle + \cdots).
$$

Agora expandimos e agrupamos os termos por potências de $\lambda$.

---

#### Ordem $\lambda^0$

$$
H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle.
$$

Esta é simplesmente a equação de autovalor do problema não perturbado.

---

#### Ordem $\lambda^1$

$$
H_0 |n^{(1)}\rangle + H_1 |n^{(0)}\rangle = E_n^{(0)} |n^{(1)}\rangle + E_n^{(1)} |n^{(0)}\rangle.
$$

Reorganizando todos os termos com $|n^{(1)}\rangle$ do mesmo lado, obtemos a **equação mestra de primeira ordem**:

$$
(H_0 - E_n^{(0)}) |n^{(1)}\rangle = - H_1 |n^{(0)}\rangle + E_n^{(1)} |n^{(0)}\rangle.
$$

---

#### Ordem $\lambda^2$

$$
H_0 |n^{(2)}\rangle + H_1 |n^{(1)}\rangle = E_n^{(0)} |n^{(2)}\rangle + E_n^{(1)} |n^{(1)}\rangle + E_n^{(2)} |n^{(0)}\rangle.
$$

Reorganizando,

$$
(H_0 - E_n^{(0)}) |n^{(2)}\rangle = - H_1 |n^{(1)}\rangle + E_n^{(1)} |n^{(1)}\rangle + E_n^{(2)} |n^{(0)}\rangle.
$$

---

#### Forma geral (ordem $\lambda^k$) – equação mestra

Para ordem genérica $\lambda^k$, o mesmo procedimento leva à forma compacta

$$
(H_0 - E_n^{(0)}) |n^{(k)}\rangle = - H_1 |n^{(k-1)}\rangle + \sum_{j=1}^{k} E_n^{(j)} |n^{(k-j)}\rangle,
$$

que é a **equação mestra** da teoria de perturbação independente do tempo: ela relaciona a correção de ordem $k$ do estado $|n^{(k)}\rangle$ com as correções de ordem mais baixa das energias e dos estados.


## 4. Ordem $\lambda^0$: recuperando o problema não perturbado

Termos de ordem $\lambda^0$ (sem $\lambda$):

$$
H_0 |n^{(0)}\rangle = E_n^{(0)} |n^{(0)}\rangle,
$$

que é exatamente a equação de autovalor do Hamiltoniano não perturbado.  
Nada novo aqui — é a base de tudo.

---



## 5. Ordem $\lambda^1$: surgem as fórmulas de primeira ordem

Agora consideramos os termos de ordem $\lambda^1$.

### 5.1. Lado esquerdo

Os termos lineares em $\lambda$ são:

- $H_0$ atuando em $\lambda |n^{(1)}\rangle \Rightarrow \lambda H_0 |n^{(1)}\rangle$,
- $\lambda H_1$ atuando em $|n^{(0)}\rangle \Rightarrow \lambda H_1 |n^{(0)}\rangle$.

Portanto, o lado esquerdo em ordem $\lambda$ é

$$
\lambda H_0 |n^{(1)}\rangle + \lambda H_1 |n^{(0)}\rangle.
$$

### 5.2. Lado direito

Do lado direito, em ordem $\lambda$, temos:

- $E_n^{(0)}$ atuando em $\lambda |n^{(1)}\rangle \Rightarrow \lambda E_n^{(0)}|n^{(1)}\rangle$,
- $\lambda E_n^{(1)}$ atuando em $|n^{(0)}\rangle \Rightarrow \lambda E_n^{(1)}|n^{(0)}\rangle$.

Portanto, o lado direito em ordem $\lambda$ é

$$
\lambda E_n^{(0)}|n^{(1)}\rangle + \lambda E_n^{(1)}|n^{(0)}\rangle.
$$

### 5.3. Igualando os termos de ordem $\lambda$

Podemos cancelar o fator $\lambda$ comum, e obtemos:

$$
H_0 |n^{(1)}\rangle + H_1 |n^{(0)}\rangle
=
E_n^{(0)}|n^{(1)}\rangle + E_n^{(1)}|n^{(0)}\rangle.
$$

Reorganizando:

$$
\bigl(H_0 - E_n^{(0)}\bigr)|n^{(1)}\rangle
=
- H_1 |n^{(0)}\rangle + E_n^{(1)}|n^{(0)}\rangle.
$$

Esta é a equação fundamental de primeira ordem.



---

## 6. Projeção na base dos autovetores de $H_0$

Vamos agora projetar essa equação na base $\{|m^{(0)}\rangle\}$ dos autovetores de $H_0$.

Multiplicando à esquerda por $\langle m^{(0)}|$:

$$
\langle m^{(0)}|\bigl(H_0 - E_n^{(0)}\bigr)|n^{(1)}\rangle
=
- \langle m^{(0)}|H_1|n^{(0)}\rangle
+
E_n^{(1)}\langle m^{(0)}|n^{(0)}\rangle.
$$

Usando que

$$
\langle m^{(0)}|H_0 = E_m^{(0)}\langle m^{(0)}|
\quad \Rightarrow \quad
\langle m^{(0)}|H_0|n^{(1)}\rangle = E_m^{(0)} \langle m^{(0)}|n^{(1)}\rangle,
$$

temos:

$$
\bigl(E_m^{(0)} - E_n^{(0)}\bigr)\langle m^{(0)}|n^{(1)}\rangle
=
- \langle m^{(0)}|H_1|n^{(0)}\rangle
+
E_n^{(1)}\delta_{mn}.
$$

Agora separamos dois casos importantes.

---

### 6.1. Caso $m = n$: correção da energia de primeira ordem

Para $m = n$, o lado esquerdo fica:

$$
\bigl(E_n^{(0)} - E_n^{(0)}\bigr)\langle n^{(0)}|n^{(1)}\rangle = 0.
$$

Logo:

$$
0 = - \langle n^{(0)}|H_1|n^{(0)}\rangle + E_n^{(1)}\delta_{nn}
= - \langle n^{(0)}|H_1|n^{(0)}\rangle + E_n^{(1)}.
$$

Portanto,

$$
\boxed{
E_n^{(1)} = \langle n^{(0)}|H_1|n^{(0)}\rangle.
}
$$

Essa é a **fórmula da correção de primeira ordem da energia** no caso não degenerado.

---

### 6.2. Caso $m \neq n$: correção do estado e surgimento dos denominadores

Para $m \neq n$, temos $\delta_{mn}=0$, e a equação fica

$$
\bigl(E_m^{(0)} - E_n^{(0)}\bigr)\langle m^{(0)}|n^{(1)}\rangle
=
- \langle m^{(0)}|H_1|n^{(0)}
