# Teoria de Perturbação Dependente do Tempo
### Interaction Picture, operador de evolução e série de Dyson

---

## 1. Motivação e setup do problema

Queremos estudar a evolução de um sistema quântico cuja dinâmica é governada por um Hamiltoniano **dependente do tempo**:

$$ H(t) = H_0 + H_1(t), $$

onde:

- $H_0$ é **independente do tempo** e resolvível exatamente (sabemos autovalores e autovetores),
- $H_1(t)$ é uma **perturbação dependente do tempo**, em geral "fraca".

Assumimos que $H_0$ tem autossistema:

$$ H_0 \lvert n \rangle = E_n \lvert n \rangle. $$

O objetivo típico da teoria de perturbação dependente do tempo é:

- Calcular **probabilidades de transição** $P_{i\to f}(t)$ entre estados de $H_0$:
  $ P_{i\to f}(t) = \big\lvert \langle f \lvert U(t) \rvert i \rangle \big\rvert^2, $
  onde $U(t)$ é o operador de evolução temporal.

---

## NOTA: Evolução temporal gerada por um Hamiltoniano independente do tempo

Mesmo quando um Hamiltoniano é **independente do tempo**, ele ainda gera **evolução temporal não–trivial**. A equação de Schrödinger no quadro de Schrödinger é:

$$
i\hbar\,\frac{d}{dt}\lvert\psi(t)\rangle = H_0 \lvert\psi(t)\rangle,
\qquad \frac{dH_0}{dt}=0.
$$

Como $H_0$ não depende explicitamente do tempo, podemos resolver a equação diretamente. Obtemos:

$$
\lvert\psi(t)\rangle = U_0(t)\lvert\psi(0)\rangle,
\qquad U_0(t) = e^{-\frac{i}{\hbar}H_0 t}.
$$

Note que “independente do tempo” **não significa** que o sistema não evolui; significa apenas que o Hamiltoniano não varia com o tempo. Ele ainda possui autovalores não nulos:

$$
H_0 \lvert n \rangle = E_n \lvert n \rangle.
$$

Assim, ao agir sobre autoestados, obtemos fases dinâmicas:

$$
U_0(t)\lvert n\rangle = e^{-\frac{i}{\hbar}E_n t}\lvert n\rangle.
$$

Portanto, a evolução temporal de um Hamiltoniano constante corresponde a uma **rotação de fase no espaço de estados**. Essa fase é fisicamente relevante, pois interfere com outras fases e é fundamental para a dinâmica quântica.

**Conclusão:**  
Energia constante gera evolução temporal (dinâmica quântica). O papel de $H_0$ é análogo ao do momento linear como gerador de deslocamentos no espaço.

---

## 2. Operador de evolução no quadro de Schrödinger

### 2) Operador de evolução completo $ U(t) $

A equação de Schrödinger no quadro de Schrödinger é:

$$
i\hbar \frac{d}{dt}\lvert \psi_S(t)\rangle = H(t)\lvert \psi_S(t)\rangle.
$$

Definimos o operador de evolução completo $U(t)$ como:

$$
\lvert \psi_S(t)\rangle = U(t)\lvert \psi_S(0)\rangle, 
\qquad U(0)=\mathbb{1}.
$$

Derivando essa expressão no tempo:

$$
\frac{d}{dt}\lvert \psi_S(t)\rangle 
= \frac{d}{dt}\Big(U(t)\lvert\psi_S(0)\rangle\Big)
= \Big(\frac{dU(t)}{dt}\Big)\lvert\psi_S(0)\rangle.
$$

Usando a equação de Schrödinger:

$$
\frac{d}{dt}\lvert \psi_S(t)\rangle 
= -\frac{i}{\hbar} H(t)\lvert \psi_S(t)\rangle
= -\frac{i}{\hbar} H(t) U(t) \lvert \psi_S(0)\rangle.
$$

Como isso deve valer para qualquer estado inicial, concluímos:

$$
i\hbar\,\frac{dU(t)}{dt} = H(t)\,U(t).
$$

Se $H$ fosse independente do tempo (e comutasse consigo mesmo em tempos diferentes), teríamos simplesmente:

$$ U(t) = e^{-\frac{i}{\hbar} H t}. $$

Mas aqui $H(t)$ depende do tempo, e em geral:

$$ [H(t), H(t')] \neq 0, $$

logo não podemos escrever $U(t)$ como uma única exponencial simples. Vamos encontrar que o operador evolução temporal neste caso é descrito pela série de Dyson:

$$
U(t) = \mathbb{1} 
+ \left(-\frac{i}{\hbar}\right) \int H(t)\, dt 
+ \left(-\frac{i}{\hbar}\right)^2 \int \!\!\int H(t)\, H(t')\, dt\, dt' 
+ \cdots
$$

---

### 2.2) Operador de evolução livre $ U_0(t) $

Agora definimos o operador de evolução **apenas** com o Hamiltoniano não‐interativo $H_0$:

$$
\lvert \psi_0(t)\rangle = U_0(t)\lvert \psi_0(0)\rangle, 
\qquad U_0(0)=\mathbb{1}.
$$

Aplicamos exatamente o mesmo procedimento. Derivando:

$$
\frac{d}{dt}\lvert \psi_0(t)\rangle 
= \Big(\frac{dU_0(t)}{dt}\Big)\lvert\psi_0(0)\rangle.
$$

Pela equação de Schrödinger, agora tomando apenas $H_0$:

$$
\frac{d}{dt}\lvert \psi_0(t)\rangle 
= -\frac{i}{\hbar} H_0\lvert \psi_0(t)\rangle
= -\frac{i}{\hbar} H_0 U_0(t) \lvert \psi_0(0)\rangle.
$$

Logo:

$$
i\hbar\,\frac{dU_0(t)}{dt} = H_0\,U_0(t).
$$

E, como $H_0$ não depende do tempo, podemos resolver explicitamente:

$$
U_0(t)=e^{-\frac{i}{\hbar}H_0 t}.
$$

---

## 3. Separando a parte "simples": definição de $H_0$ e $H_1(t)$

Escolhemos decompor:

$$ H(t) = H_0 + H_1(t), $$

onde:

- $H_0$ é a parte **não interativa** (ou resolvível),
- $H_1(t)$ é a parte **interativa/perturbativa**.

Primeiro, resolvemos a evolução associada apenas a $H_0$:

$$ i\hbar \frac{d}{dt} U_0(t) = H_0 U_0(t), \qquad U_0(0) = \mathbb{1}. $$

Como $H_0$ é independente do tempo, temos

$$ U_0(t) = e^{-\frac{i}{\hbar} H_0 t}. $$

Esse $U_0(t)$ será a peça chave da **Interaction Picture**.

---

## 4. Schrödinger vs Heisenberg vs Interaction Picture

### 4.1. Expectation values são invariantes

O valor esperado de um operador $\hat O$ em um estado $\lvert \psi(t)\rangle$ é

$$ \langle \hat O \rangle (t) = \langle \psi(t) \lvert \hat O \rvert \psi(t) \rangle. $$

Qualquer "picture" (Schrödinger, Heisenberg, interação) é apenas uma forma diferente de distribuir a dependência temporal entre $\lvert \psi \rangle$ e $\hat O$, **sem alterar** esse valor esperado.


### 4.2. Quadro de Schrödinger

- Estados evoluem: $ \lvert \psi_S(t) \rangle = U(t) \lvert \psi_S(0)\rangle. $
- Operadores (sem dependência explícita em $t$) são constantes: $O_S$ não depende de $t$.

Valor esperado:

$$ \langle \hat O \rangle (t) = \langle \psi_S(t) \lvert O_S \rvert \psi_S(t) \rangle. $$


### 4.3. Quadro de Heisenberg

- Estado é constante: $\lvert \psi_H(t) \rangle = \lvert \psi_S(0)\rangle. $
- Operador carrega toda a evolução:  

$$ O_H(t) = U^\dagger(t)\, O_S\, U(t). $$

Valor esperado:

$$ \langle \hat O \rangle (t) 
= \langle \psi_H \lvert O_H(t) \rvert \psi_H \rangle
= \langle \psi_S(t) \lvert O_S \rvert \psi_S(t) \rangle. $$


### 4.4. Quadro de interação

Não queremos nem toda a evolução no estado (Schrödinger), nem toda no operador (Heisenberg).  
Queremos separar:

- A evolução por $H_0$ será tratada como "simples" e atribuída aos operadores.
- A evolução por $H_1(t)$ será tratada como "interação" e colocada no estado.

Definimos o estado no quadro de interação por: 

$$ \lvert \psi_I(t) \rangle = U_0^\dagger(t)\, \lvert \psi_S(t) \rangle
= e^{\frac{i}{\hbar} H_0 t}\, \lvert \psi_S(t) \rangle. $$

$$\boxed{\lvert \psi_I(t) \rangle = e^{\frac{i}{\hbar} H_0 t}\, \lvert \psi_S(t) \rangle}$$

**isto é, estamos removendo a evolução temporal de $H_0$ do estado interativo**

No instante inicial:

$$ \lvert \psi_I(0) \rangle = \lvert \psi_S(0) \rangle. $$

E para um operador $A$ que não depende explicitamente de $t$ em Schrödinger:

$$ A_I(t) = U_0^\dagger(t)\, A_S\, U_0(t) 
= e^{\frac{i}{\hbar} H_0 t} A_S e^{-\frac{i}{\hbar} H_0 t}. $$


$$\boxed{A_I(t) = e^{\frac{i}{\hbar} H_0 t} A_S e^{-\frac{i}{\hbar} H_0 t}}$$
O valor esperado continua o mesmo:

$$
\langle \hat O \rangle(t) 
= \langle \psi_S(t) \lvert O_S \rvert \psi_S(t) \rangle
= \langle \psi_I(t) \lvert O_I(t) \rvert \psi_I(t) \rangle.
$$

### Resumindo: No estado de Interação os **Estados** só evoluem devido à Interação (perturbação).

## 5. Operador W(t) e o operador de evolução no quadro de interação

Definimos o operador $W(t)$ como a parte da evolução que **sobra** depois de remover $U_0(t)$:

$$
\lvert \psi_I(t) \rangle 
= U_0^\dagger(t) \lvert \psi_S(t) \rangle.
 $$

Mas:

$$
\lvert \psi_S(t) \rangle 
= U(t) \lvert \psi_S(0) \rangle.
 $$


Substituindo:

$$
\lvert \psi_I(t) \rangle 
= U_0^\dagger(t) U(t) \lvert \psi_S(0) \rangle.
$$

Mas:

$$
\lvert \psi_I(0) \rangle 
= \lvert \psi_S(0) \rangle.
$$

Então:

$$
\lvert \psi_I(t) \rangle 
= U_0^\dagger(t) U(t) \lvert \psi_I(0) \rangle.
$$

Portanto chegamos à composição de $W(t)$:

$$\boxed{ W(t) = U_0^\dagger(t) U(t). }$$

---
Sabemos que:

$$ i\hbar \frac{d}{dt} U(t) = H(t) U(t), \qquad
i\hbar \frac{d}{dt} U_0(t) = H_0 U_0(t). $$

Vamos derivar $W(t)$. Usando regra do produto:

$$
\frac{d}{dt} W(t) 
= \frac{d}{dt} \big( U_0^\dagger(t) U(t) \big)
= \big( \frac{d}{dt} U_0^\dagger(t) \big) U(t)
  + U_0^\dagger(t) \frac{d}{dt} U(t).
 $$

Da equação para $U_0(t)$:

$$ i\hbar \frac{d}{dt} U_0(t) = H_0 U_0(t)
\quad\Rightarrow\quad
- i\hbar \frac{d}{dt} U_0^\dagger(t) = U_0^\dagger(t) H_0. $$

Portanto:

$$
\frac{d}{dt} U_0^\dagger(t) = \frac{i}{\hbar} U_0^\dagger(t) H_0.
 $$

E para $U(t)$:

$$
\frac{d}{dt} U(t) = -\frac{i}{\hbar} H(t) U(t).
 $$

Substituindo em $\frac{dW}{dt}$:

$$
\frac{d}{dt} W(t)
= \frac{i}{\hbar} U_0^\dagger(t) H_0 U(t)
  -\frac{i}{\hbar} U_0^\dagger(t) H(t) U(t).
 $$

Como $H(t) = H_0 + H_1(t)$:

$$
\frac{d}{dt} W(t)
= \frac{i}{\hbar} U_0^\dagger H_0 U
  -\frac{i}{\hbar} U_0^\dagger (H_0+H_1) U
= -\frac{i}{\hbar} U_0^\dagger H_1(t) U.
 $$

Multiplicando por $i\hbar$:

$$
i\hbar \frac{d}{dt} W(t) = U_0^\dagger(t) H_1(t) U(t).
 $$

Usando $ U(t) = U_0(t) W(t) $, obtemos:

$$
i\hbar \frac{d}{dt} W(t)
= U_0^\dagger(t) H_1(t) U_0(t)\, W(t).
 $$

Definimos o Hamiltoniano de interação:

$$
H_I(t) \equiv U_0^\dagger(t) H_1(t) U_0(t).
 $$

Então a equação de evolução em Interaction Picture fica:

$$
\boxed{
i\hbar \frac{d}{dt} W(t) = H_I(t)\, W(t), \qquad W(0)=\mathbb{1}.
}
 $$

E, consequentemente, para o estado:

$$
\lvert \psi_I(t) \rangle = W(t) \lvert \psi_I(0) \rangle,
\qquad
i\hbar \frac{d}{dt} \lvert \psi_I(t) \rangle
= H_I(t) \lvert \psi_I(t) \rangle.
 $$

Agora **toda a dificuldade está em $H_1(t)$**, que aparece na forma transformada $H_I(t)$.

---


## 6. Equação integral para W(t) e série de Dyson

Da equação:

$$ i\hbar \frac{d}{dt} W(t) = H_I(t) W(t), \qquad W(0)=\mathbb{1}, $$

escrevemos:

$$
\frac{d}{dt} W(t) = -\frac{i}{\hbar} H_I(t) W(t).
 $$

Integramos de $0$ até $t$:

$$
W(t) = W(0) - \frac{i}{\hbar} \int_0^t dt' \, H_I(t') W(t').
 $$

Como $W(0)=\mathbb{1}$, temos a **equação integral**:

$$
\boxed{
W(t) = \mathbb{1} - \frac{i}{\hbar} \int_0^t dt' \, H_I(t') W(t').
}
 $$

Agora resolvemos iterativamente (método de substituição sucessiva):

1. 1ª aproximação: ignore $W(t')$ dentro da integral e substitua por $\mathbb{1}$:

$$ W(t) \approx \mathbb{1} - \frac{i}{\hbar} \int_0^t dt'_1 \, H_I(t'_1). $$

2. Substitua essa expressão de volta em $W(t')$ na integral → aparece um termo de ordem $H_I^2$:

$$
W(t) \approx \mathbb{1}
- \frac{i}{\hbar} \int_0^t dt'_1 H_I(t'_1)
+ \left( -\frac{i}{\hbar} \right)^2
  \int_0^t dt'_1 \int_0^{t'_1} dt'_2 \, H_I(t'_1) H_I(t'_2)
+ \dots
 $$

Em geral, surge a **série de Dyson**:

$$
\boxed{
W(t) = \mathbb{1}
+ \sum_{n=1}^{\infty}
\left( -\frac{i}{\hbar} \right)^n
\int_0^t dt_1 \int_0^{t_1} dt_2 \dots \int_0^{t_{n-1}} dt_n\,
H_I(t_1) H_I(t_2) \dots H_I(t_n).
}
 $$

A ordenação $0 \le t_n \le \dots \le t_2 \le t_1 \le t$ reflete a **ordenação temporal** (time ordering) implícita.

## 7. Expansão em autovetores de H₀ e coeficientes cₙ(t)

Sabemos que $H_0 \lvert n \rangle = E_n \lvert n \rangle$.  
No Interaction Picture, usamos essa base fixa e escrevemos:

$$
\lvert \psi_I(t) \rangle = \sum_n c_n(t) \lvert n \rangle.
 $$

Os coeficientes $c_n(t)$ contêm **toda a informação de transições induzidas por $H_1(t)$**.

Como:

$$
\lvert \psi_I(t) \rangle = W(t) \lvert i \rangle,
 $$

se o sistema começa no estado inicial $\lvert i \rangle$ (autoestado de $H_0$), então:

$$
c_n(t) = \langle n \lvert \psi_I(t) \rangle
= \langle n \lvert W(t) \rvert i \rangle.
 $$

A probabilidade de encontrar o sistema em $\lvert n \rangle$ no tempo $t$ é:

$$
P_{i\to n}(t) = |c_n(t)|^2.
 $$

---

## 8. Perturbação em primeira ordem

Substituindo a expansão de Dyson na expressão de $c_n(t)$, em **primeira ordem** temos:

$$
W(t) \approx \mathbb{1}
- \frac{i}{\hbar} \int_0^t dt' \, H_I(t').
 $$

Portanto,

$$
c_n^{(1)}(t)
= \langle n \lvert W(t) \rvert i \rangle_{\text{(1ª ordem)}}
= -\frac{i}{\hbar} \int_0^t dt' \, \langle n \lvert H_I(t') \rvert i \rangle.
 $$

Assim, a probabilidade de transição $i \to n$ até primeira ordem em $H_1$ é

$$ P_{i\to n}(t) \approx \big\lvert c_n^{(1)}(t) \big\rvert^2. $$

---

## 9. Perturbação em segunda ordem (esboço)

O termo de segunda ordem da série de Dyson é

$$
W^{(2)}(t)
= \left( -\frac{i}{\hbar} \right)^2
\int_0^t dt'_1 \int_0^{t'_1} dt'_2 \,
H_I(t'_1) H_I(t'_2).
 $$

Logo,

$$
c_n^{(2)}(t) 
= \langle n \lvert W^{(2)}(t) \rvert i \rangle
= \left( -\frac{i}{\hbar} \right)^2
\int_0^t dt'_1 \int_0^{t'_1} dt'_2 \,
\sum_k
\langle n \lvert H_I(t'_1) \rvert k \rangle
\langle k \lvert H_I(t'_2) \rvert i \rangle.
 $$

Aqui aparece explicitamente a **sequência de transições no tempo**:

$$
\lvert i \rangle \to \lvert k \rangle \to \lvert n \rangle,
\quad
0 \le t'_2 \le t'_1 \le t.
 $$

Essa ordenação temporal é justamente o que a série de Dyson codifica.

---

## 10. Resumo conceitual

- Escrevemos $H(t) = H_0 + H_1(t)$ e escolhemos $H_0$ como parte "resolvida".
- Definimos o **Interaction Picture** movendo a evolução de $H_0$ para os operadores:
  - $ \lvert \psi_I(t) \rangle = U_0^\dagger(t) \lvert \psi_S(t) \rangle. $
  - $ A_I(t) = U_0^\dagger(t) A_S U_0(t). $
- Definimos $ W(t) = U_0^\dagger(t) U(t) $ e mostramos que
  $ i\hbar \frac{d}{dt} W(t) = H_I(t) W(t), $
  com $ H_I(t) = U_0^\dagger(t) H_1(t) U_0(t). $
- A equação integral para $W(t)$ leva naturalmente à **série de Dyson**.
- Expandindo o estado em autoestados de $H_0$, os coeficientes $c_n(t)$ são dados por integrais dos elementos de matriz de $H_I(t)$.
- A 1ª ordem já fornece a fórmula padrão de perturbação dependente do tempo para as probabilidades de transição.
