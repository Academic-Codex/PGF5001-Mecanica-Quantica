# Visão Geral da Aula: Momento Angular Orbital e os Harmônicos Esféricos

## 1. A Motivação

Nas aulas anteriores, desenvolvemos uma teoria **abstrata** e poderosa do momento angular.  
Partindo apenas da álgebra dos geradores de rotação, $[\hat{J}_i, \hat{J}_j] = i\hbar \epsilon_{ijk}\hat{J}_k$, deduzimos que:

- Existem autoestados simultâneos de $\hat{J}^2$ e $\hat{J}_z$;
- Os autovalores são quantizados, com um número quântico $j$ que pode ser **inteiro ou semi-inteiro** ($0, 1/2, 1, 3/2, \dots$).

A pergunta que esta aula responde é:  
**O que essa teoria significa para uma partícula real, sem spin, movendo-se no espaço tridimensional?**

Estamos procurando uma **realização concreta** dessa álgebra abstrata.  
Queremos encontrar os operadores e as funções de onda que descrevem o momento angular que surge do movimento de uma partícula no espaço — o que chamamos de **momento angular orbital**, $\hat{L}$.  
O objetivo é encontrar as funções de onda $\psi_{lm}(r)$ que correspondem aos kets abstratos $|l, m\rangle$.

---

## 2. O Insight da Álgebra Linear e a Estratégia de Resolução

A estratégia aqui é uma bela aplicação do método científico dentro da matemática:

### 1. Formular uma Hipótese
Começamos com a definição clássica de momento angular orbital:
$$
\vec{L} = \vec{r} \times \vec{p},
$$
e a promovemos a operadores quânticos:
$$
\hat{L} = \hat{r} \times \hat{p}.
$$

---

### 2. Verificar a Estrutura Algébrica

O **insight da Álgebra Linear** é que, para que $\hat{L}$ seja legitimamente um momento angular, seus componentes devem satisfazer a álgebra de Lie das rotações.  
Portanto, o primeiro passo crucial é verificar se nossa definição satisfaz as relações de comutação fundamentais:

$$
[\hat{L}_i, \hat{L}_j] = i\hbar \epsilon_{ijk} \hat{L}_k.
$$

Se isso for verdade, então toda a poderosa maquinaria da teoria do momento angular (autovalores, operadores escada, etc.) aplica-se diretamente a $\hat{L}$.

---

### 3. Resolver as Equações de Autovalor

Uma vez confirmada a álgebra, o problema se resume a resolver as equações de autovalor na base de posição (isto é, como equações diferenciais para a função de onda $\psi(r) = \langle r|\psi\rangle$):

$$
\hat{L}^2 \psi_{lm}(r) = \hbar^2 l(l+1)\psi_{lm}(r)
$$
$$
\hat{L}_z \psi_{lm}(r) = \hbar m \psi_{lm}(r)
$$

Como o problema tem **simetria esférica**, a abordagem mais inteligente é expressar os operadores $\hat{L}_i$ em coordenadas esféricas $(r, \theta, \phi)$, o que simplificará enormemente a resolução.

---

## 3. Os Resultados Principais da Aula

- **A Definição de $\hat{L}$ é Válida:**  
  A primeira grande conclusão é que a definição $\hat{L} = \hat{r} \times \hat{p}$ de fato satisfaz a álgebra do momento angular.  
  Isso solidifica a conexão entre a definição clássica e a estrutura quântica das rotações.

---

- **Quantização do Momento Angular Orbital é Inteira:**  
  Este é talvez o resultado mais profundo da aula.  
  Ao resolver a equação de autovalor para $\hat{L}_z$, encontramos que a dependência azimutal ($\phi$) da função de onda é da forma $e^{im\phi}$.

  A condição física de que a função de onda deve ser **unívoca** (ter um único valor em cada ponto do espaço) exige que:
  $$
  \psi(\phi) = \psi(\phi + 2\pi)
  $$

  Isso impõe a restrição de que $m$ (e, consequentemente, $l$) deve ser um **número inteiro**.  
  O momento angular orbital, associado ao movimento no espaço, não pode ser semi-inteiro — este tipo de quantização (como o **spin**) é uma propriedade intrínseca da partícula, não espacial.

---

- **Os Operadores em Coordenadas Esféricas:**  
  Encontramos as expressões explícitas para os operadores $\hat{L}_z$, $\hat{L}_\pm$ e $\hat{L}^2$ em coordenadas esféricas.  
  Notavelmente, $\hat{L}^2$ corresponde exatamente à parte angular do operador Laplaciano, reforçando a conexão entre momento angular e a energia cinética de rotação.

---

- **Os Harmônicos Esféricos $Y_{lm}(\theta, \phi)$:**  
  As soluções das equações de autovalor simultâneas para $\hat{L}^2$ e $\hat{L}_z$ são as funções chamadas **Harmônicos Esféricos** $Y_{lm}(\theta, \phi)$.

  Estas funções são a **representação na base de posição angular** $\langle \theta, \phi|$ do estado abstrato $|l, m\rangle$.  
  Elas formam uma base ortonormal para qualquer função definida na superfície de uma esfera.

---

- **Relação com Polinômios de Legendre:**  
  A derivação sistemática dos $Y_{lm}$ revela sua estrutura interna.  
  Eles são construídos a partir de:

  - **Polinômios de Legendre:** $P_l(\cos\theta)$ — para o caso $m = 0$.
  - **Funções Associadas de Legendre:** $P_l^m(\cos\theta)$ — para $m \neq 0$.

  A derivação segue uma estratégia elegante:  
  encontra-se primeiro o estado com $m$ máximo ($m = l$), o “estado esticado” $|l, l\rangle$, e aplica-se sucessivamente o operador de abaixamento $\hat{L}_-$ para gerar todos os outros estados $|l, m\rangle$ com $m < l$.

---

## Conclusão

Nesta aula, vimos como o momento angular orbital é uma **realização concreta** da álgebra de rotações.  
A teoria abstrata dos operadores $\hat{J}$ ganha forma física no espaço tridimensional por meio das funções $Y_{lm}(\theta,\phi)$ — os harmônicos esféricos — que descrevem a dependência angular de todas as funções de onda com simetria esférica.

Eles constituem uma das bases matemáticas mais importantes da física teórica, aparecendo em:

- Estrutura eletrônica de átomos;
- Expansão multipolar em eletrodinâmica;
- Ondas em superfícies esféricas (acústicas, gravitacionais, quânticas);
- E em toda formulação moderna de simetrias contínuas.

# Derivação Matemática

# 1. Verificação da Álgebra do Momento Angular Orbital

**Motivação.** O primeiro passo é provar que a definição quântica emprestada da mecânica clássica,
$\hat{L}=\hat{r}\times\hat{p}$, é consistente com a teoria abstrata do momento angular.  
Devemos mostrar a álgebra de Lie: $[\hat{L}_i,\hat{L}_j]=i\hbar\,\epsilon_{ijk}\hat{L}_k$.  
Aqui provamos explicitamente $[\hat{L}_x,\hat{L}_y]=i\hbar\hat{L}_z$.

## Derivação passo a passo

### 1) Definir os operadores (forma cartesiana)
$$
\hat{L}_x=\hat{y}\hat{p}_z-\hat{z}\hat{p}_y,\qquad
\hat{L}_y=\hat{z}\hat{p}_x-\hat{x}\hat{p}_z,\qquad
\hat{L}_z=\hat{x}\hat{p}_y-\hat{y}\hat{p}_x.
$$

### 2) Escrever o comutador
$$
[\hat{L}_x,\hat{L}_y]=[\hat{y}\hat{p}_z-\hat{z}\hat{p}_y,\;\hat{z}\hat{p}_x-\hat{x}\hat{p}_z].
$$

**Insight.** Usaremos distributividade de comutadores,  
$[A-B,C-D]=[A,C]-[A,D]-[B,C]+[B,D]$.

### 3) Expandir
$$
[\hat{L}_x,\hat{L}_y]
=
[\hat{y}\hat{p}_z,\hat{z}\hat{p}_x]
-
[\hat{y}\hat{p}_z,\hat{x}\hat{p}_z]
-
[\hat{z}\hat{p}_y,\hat{z}\hat{p}_x]
+
[\hat{z}\hat{p}_y,\hat{x}\hat{p}_z].
$$

### 4) Resolver termo a termo
Usamos $[AB,C]=A[B,C]+[A,C]B$ e as CCR:  
$[\hat{x}_i,\hat{p}_j]=i\hbar\delta_{ij}$, enquanto operadores de partículas/eixos distintos comutam.

- **(1)** $[\hat{y}\hat{p}_z,\hat{z}\hat{p}_x]
= \hat{y}[\hat{p}_z,\hat{z}]\hat{p}_x + [\hat{y},\hat{z}]\hat{p}_x\hat{p}_z + \hat{z}[\hat{y},\hat{p}_x]\hat{p}_z
= -i\hbar\,\hat{y}\hat{p}_x + 0 + 0
= -i\hbar\,\hat{y}\hat{p}_x.$

- **(2)** $[\hat{y}\hat{p}_z,\hat{x}\hat{p}_z]=0$ (todos os fatores comutam).

- **(3)** $[\hat{z}\hat{p}_y,\hat{z}\hat{p}_x]=0$ (mesmo motivo).

- **(4)** $[\hat{z}\hat{p}_y,\hat{x}\hat{p}_z]
= \hat{z}[\hat{p}_y,\hat{x}]\hat{p}_z + [\hat{z},\hat{x}]\hat{p}_z\hat{p}_y + \hat{x}[\hat{z},\hat{p}_z]\hat{p}_y
= 0 + 0 + i\hbar\,\hat{x}\hat{p}_y
= i\hbar\,\hat{x}\hat{p}_y.$

### 5) Combinar
$$
[\hat{L}_x,\hat{L}_y] = (-i\hbar\,\hat{y}\hat{p}_x) + 0 + 0 + (i\hbar\,\hat{x}\hat{p}_y)
= i\hbar(\hat{x}\hat{p}_y - \hat{y}\hat{p}_x)
= i\hbar\,\hat{L}_z.
$$

### 6) Conclusão
$$
[\hat{L}_x,\hat{L}_y]=i\hbar\,\hat{L}_z,
$$
e, por permutações cíclicas, obtemos toda a álgebra do momento angular.

---

# 2. Operadores de Momento Angular em Coordenadas Esféricas

**Motivação.** Para resolver autovalores, é natural trabalhar em um sistema que respeite a simetria esférica.

## 2.1 Relações de coordenadas
$$
x=r\sin\theta\cos\phi,\qquad
y=r\sin\theta\sin\phi,\qquad
z=r\cos\theta.
$$

## 2.2 Regra da cadeia para $\partial/\partial\phi$
$$
\frac{\partial}{\partial\phi}
= \frac{\partial x}{\partial\phi}\frac{\partial}{\partial x}
+ \frac{\partial y}{\partial\phi}\frac{\partial}{\partial y}
+ \frac{\partial z}{\partial\phi}\frac{\partial}{\partial z}
= (-y)\frac{\partial}{\partial x} + x\frac{\partial}{\partial y},
$$
pois $\frac{\partial x}{\partial\phi}=-r\sin\theta\sin\phi=-y$,  
$\frac{\partial y}{\partial\phi}=r\sin\theta\cos\phi=x$,  
$\frac{\partial z}{\partial\phi}=0$.

## 2.3 Derivação de $\hat{L}_z$
Com $\hat{p}_i=-i\hbar\,\partial/\partial x_i$,
$$
\hat{L}_z=\hat{x}\hat{p}_y-\hat{y}\hat{p}_x
= x(-i\hbar\frac{\partial}{\partial y})
- y(-i\hbar\frac{\partial}{\partial x})
= -i\hbar\left(x\frac{\partial}{\partial y}-y\frac{\partial}{\partial x}\right)
= -i\hbar\frac{\partial}{\partial\phi}.
$$

## 2.4 Expressões de $\hat{L}_x$ e $\hat{L}_y$ (e escada)
O resultado padrão (obtido via inversão do Jacobiano ou via operadores escada) é:
$$
\hat{L}_x
= i\hbar\!\left(\sin\phi\,\frac{\partial}{\partial\theta}
+ \cot\theta\,\cos\phi\,\frac{\partial}{\partial\phi}\right),\qquad
\hat{L}_y
= i\hbar\!\left(-\cos\phi\,\frac{\partial}{\partial\theta}
+ \cot\theta\,\sin\phi\,\frac{\partial}{\partial\phi}\right).
$$

Combinando,
$$
\hat{L}_\pm=\hat{L}_x\pm i\hat{L}_y
= \hbar\,e^{\pm i\phi}\!\left(\pm\,\frac{\partial}{\partial\theta}
+ i\cot\theta\,\frac{\partial}{\partial\phi}\right).
$$

## 2.5 Operador $\hat{L}^2$ (forma esférica)
Usando a identidade
$$
\hat{L}^2 = \tfrac12(\hat{L}_+\hat{L}_-+\hat{L}_-\hat{L}_+) + \hat{L}_z^2,
$$
chega-se à forma diferencial puramente angular:
$$
\boxed{
\hat{L}^2
= -\hbar^2\!\left[
\frac{1}{\sin\theta}\frac{\partial}{\partial\theta}
\Big(\sin\theta\,\frac{\partial}{\partial\theta}\Big)
+ \frac{1}{\sin^2\theta}\frac{\partial^2}{\partial\phi^2}
\right].
}
$$

Essa é exatamente a parte angular do Laplaciano em coordenadas esféricas e leva aos **harmônicos esféricos** $Y_{lm}(\theta,\phi)$ como autoestados simultâneos de $\hat{L}^2$ e $\hat{L}_z$:
$$
\hat{L}^2 Y_{lm}=\hbar^2 l(l+1)Y_{lm},\qquad
\hat{L}_z Y_{lm}=\hbar m\,Y_{lm}.
$$

# 3. Expansão de $\hat{L}_-\hat{L}_+$ e Dedução Final de $\hat{L}^2$

A expansão do produto dos operadores escada $\hat{L}_-\hat{L}_+$ começa substituindo suas definições em coordenadas esféricas:

$$
\hat{L}_-\hat{L}_+ =
\Big[\hbar e^{-i\phi}\!\left(-\frac{\partial}{\partial\theta} + i\cot\theta\,\frac{\partial}{\partial\phi}\right)\!\Big]
\Big[\hbar e^{i\phi}\!\left(\frac{\partial}{\partial\theta} + i\cot\theta\,\frac{\partial}{\partial\phi}\right)\!\Big]
$$

$$
= \hbar^2 e^{-i\phi}
\!\left(-\frac{\partial}{\partial\theta} + i\cot\theta\,\frac{\partial}{\partial\phi}\right)\!
\left(e^{i\phi}\frac{\partial}{\partial\theta} + i e^{i\phi}\cot\theta\,\frac{\partial}{\partial\phi}\right)
$$

Aplicando a regra do produto (lembrando que $\frac{\partial e^{i\phi}}{\partial\theta}=0$, mas $\frac{\partial e^{i\phi}}{\partial\phi}=i e^{i\phi}$):

$$
= \hbar^2 e^{-i\phi}\Bigg[
- \frac{\partial}{\partial\theta}\!\left(e^{i\phi}\frac{\partial}{\partial\theta}\right)
- \frac{\partial}{\partial\theta}\!\left(i e^{i\phi}\cot\theta\,\frac{\partial}{\partial\phi}\right)
+ i\cot\theta\,\frac{\partial}{\partial\phi}\!\left(e^{i\phi}\frac{\partial}{\partial\theta}\right)
+ (i\cot\theta)^2\frac{\partial}{\partial\phi}\!\left(e^{i\phi}\frac{\partial}{\partial\phi}\right)
\Bigg].
$$

Depois de expandir e simplificar os termos exponenciais, obtemos:

$$
\hat{L}_-\hat{L}_+ = -\hbar^2\!\left(
\frac{\partial^2}{\partial\theta^2}
+ \cot\theta\,\frac{\partial}{\partial\theta}
- 2i\cot\theta\,\frac{\partial^2}{\partial\theta\partial\phi}
+ \cot^2\theta\!\left(i\,\frac{\partial}{\partial\phi} - \frac{\partial^2}{\partial\phi^2}\right)
\right).
$$

Somando $\hat{L}_z^2 + \hbar\hat{L}_z = -\hbar^2\frac{\partial^2}{\partial\phi^2} - i\hbar^2\frac{\partial}{\partial\phi}$ e simplificando:

$$
\boxed{
\hat{L}^2 = -\hbar^2
\left[
\frac{1}{\sin\theta}\frac{\partial}{\partial\theta}
\left(\sin\theta\,\frac{\partial}{\partial\theta}\right)
+ \frac{1}{\sin^2\theta}\frac{\partial^2}{\partial\phi^2}
\right].
}
$$

Este é exatamente $-\hbar^2$ vezes a parte angular do Laplaciano.

# 4. Encontrando os Harmônicos Esféricos $Y_{lm}(\theta,\phi)$

**Motivação:** Agora que temos os operadores $\hat{L}^2$ e $\hat{L}_z$, resolvemos as equações de autovalor para encontrar suas autofunções, os **Harmônicos Esféricos** $Y_{lm}(\theta,\phi)$.

---

## Derivação passo a passo

### 1) Condições para o estado de $m$ máximo ($m = l$)

O estado $Y_{ll}$ deve satisfazer:

a) $\hat{L}_z Y_{ll} = \hbar l\, Y_{ll}$  
b) $\hat{L}_+ Y_{ll} = 0$  (não pode ser elevado mais)

---

### 2) Resolvendo $\hat{L}_z Y_{ll} = \hbar l Y_{ll}$

$$
-i\hbar \frac{\partial Y_{ll}}{\partial\phi} = \hbar l Y_{ll}
\;\Rightarrow\;
\frac{\partial Y_{ll}}{\partial\phi} = i l Y_{ll}.
$$

A solução é:
$$
Y_{ll}(\theta,\phi) = F(\theta)e^{i l\phi},
$$
onde $F(\theta)$ é uma função apenas de $\theta$.

**Condição de unicidade:**
$$
\psi(\phi + 2\pi) = \psi(\phi) \;\Rightarrow\; e^{i2\pi l}=1,
$$
portanto $l$ deve ser **inteiro**.  
O momento angular orbital é quantizado em múltiplos inteiros de $\hbar$.

---

### 3) Aplicando $\hat{L}_+ Y_{ll}=0$

$$
\hbar e^{i\phi}\!\left(\frac{\partial}{\partial\theta} + i\cot\theta\,\frac{\partial}{\partial\phi}\right)[F(\theta)e^{il\phi}] = 0
$$

Dividindo por $\hbar e^{i\phi}$ e substituindo $\frac{\partial}{\partial\phi}(F e^{il\phi}) = i l F e^{il\phi}$:

$$
\frac{dF}{d\theta} - l\cot\theta\,F = 0
\;\Rightarrow\;
\frac{dF}{F} = l \frac{\cos\theta}{\sin\theta}d\theta
\;\Rightarrow\;
F(\theta) = C(\sin\theta)^l.
$$

---

### 4) Solução final para $Y_{ll}$

$$
Y_{ll}(\theta,\phi) = C_l (\sin\theta)^l e^{i l\phi},
$$
onde $C_l$ é determinado pela normalização:
$$
\int |Y_{ll}|^2 d\Omega = 1.
$$

---

### 5) Gerando os outros $Y_{lm}$

Os demais harmônicos $Y_{lm}$ (com $m<l$) são obtidos aplicando repetidamente o operador de abaixamento $\hat{L}_-$:

$$
Y_{l,m-1} = \frac{1}{\hbar\sqrt{(l+m)(l-m+1)}} \hat{L}_- Y_{lm}.
$$

Essa operação gera as funções que envolvem os **Polinômios Associados de Legendre** $P_l^m(\cos\theta)$, levando à forma geral:

$$
Y_{lm}(\theta,\phi) =
N_{lm} P_l^m(\cos\theta)e^{i m\phi},
$$
onde $N_{lm}$ é o fator de normalização.

---

# 5. Conexão com as Matrizes D de Wigner (Tópico Avançado)

**Motivação:** Existe uma relação elegante entre os Harmônicos Esféricos e as Matrizes D de Wigner $D^l_{m'm}(R)$, que representam rotações no espaço de momento angular.

---

## 1) Propriedade de Transformação

O harmônico esférico é a função de onda do estado $|\hat{r}\rangle = |\theta,\phi\rangle$:
$$
Y_{lm}(\hat{r}) = \langle \hat{r}|l,m\rangle.
$$

Sob uma rotação $U(R)$:
$$
(U(R)Y_{lm})(\hat{r}) = Y_{lm}(R^{-1}\hat{r}).
$$

Usando a identidade de resolução da unidade $\sum_{m'} |l,m'\rangle\langle l,m'|$:
$$
(U(R)Y_{lm})(\hat{r}) =
\sum_{m'} \langle \hat{r}|l,m'\rangle \langle l,m'|U(R)|l,m\rangle
= \sum_{m'} Y_{lm'}(\hat{r}) D^l_{m'm}(R),
$$

portanto:
$$
Y_{lm}(R^{-1}\hat{r}) = \sum_{m'} Y_{lm'}(\hat{r}) D^l_{m'm}(R).
$$

---

## 2) Escolha Estratégica

Tomando $\hat{r}=(\theta,\phi)$ e escolhendo $R(\phi,\theta,0)$ (rotação que leva $\hat{z}$ até $\hat{r}$):

$$
Y_{lm}(\hat{z}) = \sum_{m'} Y_{lm'}(\theta,\phi) D^l_{m'm}(R(\phi,\theta,0)).
$$

---

## 3) Avaliando no Polo Norte

Para $\theta=0$:
$$
Y_{lm}(\theta=0) = \sqrt{\frac{2l+1}{4\pi}}\,\delta_{m0}.
$$

---

## 4) Invertendo a Relação

Substituindo esse resultado de volta:
$$
\sqrt{\frac{2l+1}{4\pi}}\delta_{m,0} = \sum_{m'} Y_{lm'}(\theta,\phi) D^l_{m'0}(R(\phi,\theta,0)),
$$

e usando a unitariedade das matrizes de Wigner:
$$
\boxed{
Y_{lm}(\theta,\phi) =
\sqrt{\frac{2l+1}{4\pi}}\, D^{l*}_{m0}(\phi,\theta,0).
}
$$

---

**Conclusão:**
Os **Harmônicos Esféricos** são, essencialmente, os elementos da **primeira coluna ($m=0$)** das Matrizes de Rotação de Wigner, **complexo conjugados**:
$$
Y_{lm}(\theta,\phi) \propto D^{l*}_{m0}(\phi,\theta,0).
$$
