# Visão Geral da Aula: Potenciais Centrais e o Problema de Dois Corpos

# Parte I - Potencial Central

## 1. A Motivação

O objetivo central da aula de hoje é desenvolver as ferramentas para resolver um dos problemas mais importantes e onipresentes da física: o movimento de uma partícula sob a influência de um **potencial central**.  
Um potencial central, $V(r)$, é aquele que depende apenas da **distância** da partícula à origem, ou seja,

$$
V(r), \quad \text{onde } r = |\vec{r}|.
$$

Por que isso é tão crucial? Porque os sistemas físicos mais fundamentais são descritos por potenciais centrais:

- **O Átomo de Hidrogênio:** a força de Coulomb entre o próton e o elétron depende de $1/r$.
- **Forças Nucleares (aproximadamente):** o potencial de Yukawa que descreve a interação entre núcleos é um potencial central.
- **Gravitação Newtoniana:** a força gravitacional entre dois corpos depende de $1/r$.

A beleza de um potencial central está em sua **simetria esférica**.  
Em mecânica quântica, simetria não é apenas um conceito estético; ela tem consequências matemáticas profundas e poderosas.  
Uma simetria contínua implica a existência de uma quantidade conservada.  
A simetria sob rotações implica a conservação do **momento angular**.

---

## 2. O Insight da Álgebra Linear

Aqui entra a conexão fundamental com a Álgebra Linear.  
Dizer que o momento angular $\hat{L}$ é conservado é o mesmo que dizer que o operador Hamiltoniano $\hat{H}$ comuta com o operador de momento angular:

$$
[\hat{H}, \hat{L}] = 0
$$

Um teorema fundamental da Álgebra Linear nos diz que **se dois operadores comutam, eles possuem um conjunto comum de autoestados (ou autovetores)**.

Isso significa que podemos encontrar estados quânticos $|\psi\rangle$ que são simultaneamente autoestados de $\hat{H}$, de $\hat{L}^2$ (o quadrado do momento angular total) e de uma de suas componentes, convencionalmente $\hat{L}_z$:

$$
\begin{cases}
\hat{H}|\psi\rangle = E|\psi\rangle \\
\hat{L}^2|\psi\rangle = \hbar^2 l(l+1)|\psi\rangle \\
\hat{L}_z|\psi\rangle = \hbar m_l|\psi\rangle
\end{cases}
$$

Essa é a chave de tudo!  
Em vez de resolver uma equação diferencial parcial complicada em 3D, podemos usar essa propriedade para **separar as variáveis radiais das angulares**.

---

## 3. Os Resultados Principais

- **Separação de Variáveis:**  
  A autofunção (a representação do autoestado $|\psi\rangle$ na base de posição) pode ser escrita como o produto de uma parte radial e uma parte angular:

  $$
  \psi(r, \theta, \phi) = R(r) Y_{lm}(\theta, \phi)
  $$

  onde os $Y_{lm}$ são os **Harmônicos Esféricos**, que já são os autoestados da parte angular.

- **Equação Radial:**  
  Ao fazer essa separação, a complexa equação de Schrödinger em 3D se reduz a uma equação diferencial ordinária muito mais simples, apenas para a parte radial $R(r)$.  
  Essa equação se parece com a equação de Schrödinger em 1D com um **potencial efetivo**

  $$
  U_{\text{eff}}(r) = V(r) + \frac{\hbar^2 l(l+1)}{2mr^2}.
  $$

  O segundo termo é a famosa **barreira centrífuga**, uma repulsão puramente quântica que afasta partículas com momento angular ($l > 0$) da origem.

- **Soluções para a Partícula Livre ($V(r)=0$):**  
  No caso de uma partícula livre, as soluções para a equação radial são as **Funções de Bessel Esféricas**.  
  A condição física de que a função de onda não pode divergir na origem nos força a descartar um conjunto de soluções (as funções de Neumann esféricas) e manter apenas as **funções de Bessel esféricas regulares**, $j_l(kr)$.

---

- **Redução do Problema de Dois Corpos:**  
  Finalmente, mostramos que o problema (aparentemente complexo) de dois corpos interagindo através de um potencial central $V(|\vec{r}_1 - \vec{r}_2|)$ pode ser exatamente separado em dois problemas de um corpo:

  1. O movimento livre do centro de massa do sistema.  
  2. O movimento de uma partícula "fictícia" com **massa reduzida** $\mu$ sob o mesmo potencial central $V(r)$.

  Isso significa que todo o formalismo que desenvolvemos para um corpo em um potencial central pode ser diretamente aplicado para resolver o problema de dois corpos, como o átomo de hidrogênio.

---

# Derivação Detalhada: Da Equação de Schrödinger 3D à Solução da Partícula Livre

Agora, vamos abrir as contas, passo a passo, como prometido.

## Motivação: Encontrando os Estados Estacionários em um Potencial Central

Nosso ponto de partida é a equação de Schrödinger independente do tempo para uma partícula de massa $m$ em um potencial central $V(r)$:

$$
\hat{H}|\psi\rangle = E|\psi\rangle
$$

### Passo 1: Escrevendo o Hamiltoniano na Base de Posição

Na representação de coordenadas (notação de função de onda), o operador $\hat{H}$ se torna:

$$
\hat{H} = \frac{\hat{p}^2}{2m} + V(r) = -\frac{\hbar^2}{2m}\nabla^2 + V(r)
$$

A equação de Schrödinger se torna uma equação diferencial parcial:

$$
\left(-\frac{\hbar^2}{2m}\nabla^2 + V(r)\right)\psi(r) = E\psi(r)
$$

---

### Passo 2: O Laplaciano em Coordenadas Esféricas e a Conexão com Momento Angular

Devido à simetria do potencial, é natural trabalhar em coordenadas esféricas $(r, \theta, \phi)$.  
O operador Laplaciano $\nabla^2$ nesta base é:

$$
\nabla^2 = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2 \frac{\partial}{\partial r}\right)
+ \frac{1}{r^2 \sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta \frac{\partial}{\partial \theta}\right)
+ \frac{1}{r^2 \sin^2\theta}\frac{\partial^2}{\partial \phi^2}
$$

Agora, o **insight crucial**: a parte angular do Laplaciano está diretamente relacionada ao operador quadrado do momento angular $\hat{L}^2$.  
Lembre-se que:

$$
\hat{L}^2 = -\hbar^2\left[\frac{1}{\sin\theta}\frac{\partial}{\partial\theta}
\left(\sin\theta \frac{\partial}{\partial\theta}\right)
+ \frac{1}{\sin^2\theta}\frac{\partial^2}{\partial\phi^2}\right]
$$

Portanto, podemos reescrever o Laplaciano de forma mais compacta e fisicamente significativa:

$$
\nabla^2 = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2 \frac{\partial}{\partial r}\right)
- \frac{\hat{L}^2}{\hbar^2 r^2}
$$

Substituindo isso na equação de Schrödinger:

$$
\left[-\frac{\hbar^2}{2m}\left(\frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial}{\partial r}\right)\right)
+ \frac{\hat{L}^2}{2mr^2} + V(r)\right]\psi(r,\theta,\phi) = E\psi(r,\theta,\phi)
$$

---

### Passo 3: Separação de Variáveis (O "Ansatz")

Aqui usamos o insight da Álgebra Linear.  
Como $[\hat{H}, \hat{L}^2] = 0$ e $[\hat{H}, \hat{L}_z] = 0$, procuramos uma solução que seja um autoestado simultâneo.  
Já conhecemos os autoestados de $\hat{L}^2$ e $\hat{L}_z$: são os **Harmônicos Esféricos**, $Y_{lm}(\theta,\phi)$.

Propomos então uma solução (um *ansatz*) na forma:

$$
\psi(r,\theta,\phi) = R_{nl}(r) Y_{lm}(\theta,\phi)
$$

Substituímos este ansatz na equação. Como $Y_{lm}$ é um autoestado de $\hat{L}^2$, temos:

$$
\hat{L}^2(R(r)Y_{lm}) = R(r)\hat{L}^2Y_{lm} = R(r)\hbar^2 l(l+1)Y_{lm}
$$

A equação de Schrödinger se torna:

$$
\left[-\frac{\hbar^2}{2m r^2}\frac{d}{dr}\left(r^2\frac{dR}{dr}\right)
+ \frac{\hbar^2 l(l+1)}{2mr^2} + V(r)\right]R(r)Y_{lm}(\theta,\phi)
= E R(r)Y_{lm}(\theta,\phi)
$$

Como o operador na esquerda não age mais nas variáveis angulares, podemos dividir ambos os lados por $Y_{lm}(\theta,\phi)$.

---

### Passo 4: A Equação Radial e o Potencial Efetivo

O que resta é a **Equação Diferencial Radial**:

$$
\left[-\frac{\hbar^2}{2m r^2}\frac{d}{dr}\left(r^2\frac{dR}{dr}\right)
+ \frac{\hbar^2 l(l+1)}{2mr^2} + V(r)\right]R(r) = E R(r)
$$

Definimos uma nova função radial $u(r) = rR(r)$.  
As derivadas são:

$$
R(r) = \frac{u(r)}{r}
$$

$$
\frac{dR}{dr} = \frac{1}{r}\frac{du}{dr} - \frac{u(r)}{r^2}
$$

Substituindo na equação radial, obtemos após simplificação:

$$
-\frac{\hbar^2}{2m}\frac{d^2u(r)}{dr^2}
+ \left[V(r) + \frac{\hbar^2 l(l+1)}{2mr^2}\right]u(r)
= E u(r)
$$

Esta é a equação de Schrödinger 1D com um **potencial efetivo**:

$$
U_{\text{eff}}(r) = V(r) + \frac{\hbar^2 l(l+1)}{2mr^2}
$$

---

### Caso Especial: A Partícula Livre ($V(r) = 0$)

Para $V(r)=0$, a equação radial fica:

$$
-\frac{\hbar^2}{2m}\frac{d^2u}{dr^2}
+ \frac{\hbar^2 l(l+1)}{2mr^2}u(r)
= E u(r)
$$

Com $E = \frac{\hbar^2 k^2}{2m}$, obtemos:

$$
\frac{d^2u}{dr^2}
+ \left(k^2 - \frac{l(l+1)}{r^2}\right)u(r) = 0
$$

Esta é a **Equação Diferencial de Bessel Esférica**.

---

### Passo 5: Aplicando Condições Físicas

As soluções são:

$$
R(r) = A j_l(kr) + B n_l(kr)
$$

- As **Funções de Bessel Esféricas** $j_l(kr)$ são regulares na origem.
- As **Funções de Neumann Esféricas** $n_l(kr)$ divergem em $r=0$.

Portanto, impondo a condição física de normalização, obtemos $B=0$ e:

$$
R_k(r) = A_k j_l(kr)
$$

A solução completa é:

$$
\psi_{klm}(r,\theta,\phi) = A_k j_l(kr) Y_{lm}(\theta,\phi)
$$

Estas são as **ondas esféricas** da mecânica quântica, em contraste com as ondas planas $e^{i\vec{k}\cdot\vec{r}}$, que são autoestados do operador momento linear $\hat{p}$.

Ambas formam uma base completa para descrever uma partícula livre.

# Parte II

# O Problema de Dois Corpos e a Simetria

## 1. A Motivação

Até agora, tratamos de uma única partícula em um potencial central.  
No entanto, a maioria dos sistemas realistas — como o átomo de hidrogênio (próton + elétron) ou um sistema planetário (estrela + planeta) — envolve **dois corpos interagindo**.

O Hamiltoniano para tal sistema parece consideravelmente mais complicado, envolvendo as coordenadas e momentos de **ambas as partículas**.

A pergunta central é:  
**será que podemos simplificar este problema de seis graus de liberdade**  
$(x_1, y_1, z_1, x_2, y_2, z_2)$  
**para algo que já sabemos resolver?**

A resposta é um retumbante **sim**, e a chave, mais uma vez, está na **simetria**.  
Se o potencial de interação depende apenas da **distância relativa** entre as partículas, $V(|\vec{r}_1 - \vec{r}_2|)$, o sistema como um todo possui **duas simetrias fundamentais**:

1. **Invariância Translacional:**  
   O sistema se comporta da mesma forma se movermos ambos os corpos pela mesma quantidade no espaço.  
   Isso implica a conservação do **momento linear total**.

2. **Invariância Rotacional:**  
   O sistema se comporta da mesma forma se girarmos ambos os corpos pelo mesmo ângulo em torno de uma origem comum.  
   Isso implica a conservação do **momento angular total**.

Nosso objetivo é usar essas simetrias para fazer uma **mudança de coordenadas** que separe o Hamiltoniano em partes mais simples e independentes.  
Esta não é uma aproximação — é uma **decomposição exata** do problema.

---

## 2. O Hamiltoniano de Dois Corpos

O Hamiltoniano clássico para duas partículas de massas $m_1$ e $m_2$, posições $\vec{r}_1, \vec{r}_2$ e momentos $\vec{p}_1, \vec{p}_2$, interagindo por um potencial $V(|\vec{r}_1 - \vec{r}_2|)$ é:

$$
H = \frac{\vec{p}_1^2}{2m_1} + \frac{\vec{p}_2^2}{2m_2} + V(|\vec{r}_1 - \vec{r}_2|)
$$

---

## 3. Mudança para Coordenadas Mais Naturais

Introduzimos duas novas coordenadas:

- **Centro de Massa (CM):**
  $$
  \vec{R} = \frac{m_1 \vec{r}_1 + m_2 \vec{r}_2}{m_1 + m_2}
  $$

- **Coordenada Relativa:**
  $$
  \vec{r} = \vec{r}_1 - \vec{r}_2
  $$

Os momentos conjugados correspondentes são:

$$
\vec{P} = \vec{p}_1 + \vec{p}_2
\quad\text{e}\quad
\vec{p} = \frac{m_2 \vec{p}_1 - m_1 \vec{p}_2}{m_1 + m_2}
$$

onde $\vec{P}$ é o **momento linear total** e $\vec{p}$ é o **momento relativo**.

---

## 4. O Resultado: Separação Exata do Hamiltoniano

Substituindo essas definições no Hamiltoniano, obtemos:

$$
H = \frac{\vec{P}^2}{2M} + \frac{\vec{p}^2}{2\mu} + V(r)
$$

onde:
- $M = m_1 + m_2$ é a **massa total**,
- $\mu = \frac{m_1 m_2}{m_1 + m_2}$ é a **massa reduzida**.

Assim, o problema se divide elegantemente em **duas partes independentes**:

1. **Movimento do Centro de Massa:**
   $$
   H_{CM} = \frac{\vec{P}^2}{2M}
   $$
   Representa o movimento livre do sistema como um todo.

2. **Movimento Relativo:**
   $$
   H_{rel} = \frac{\vec{p}^2}{2\mu} + V(r)
   $$
   Representa o movimento de uma partícula "fictícia" de massa $\mu$ em um potencial central $V(r)$.

---

## 5. Interpretação Física

Isso significa que **todo o problema de dois corpos** pode ser tratado como **um único corpo** de massa $\mu$ movendo-se em torno de um ponto fixo sob o mesmo potencial $V(r)$.

- No átomo de hidrogênio, a massa reduzida é:
  $$
  \mu = \frac{m_p m_e}{m_p + m_e} \approx m_e
  $$
  já que $m_p \gg m_e$.

Portanto, todo o formalismo que desenvolvemos para o **problema de um corpo em um potencial central** aplica-se diretamente ao **problema de dois corpos**.

---

## 6. Conclusão

Através da **simetria** e da **mudança de coordenadas**, conseguimos:

- Separar o movimento coletivo (centro de massa) do movimento relativo;
- Reduzir o problema de dois corpos a um problema de um corpo;
- Preservar a forma da equação de Schrödinger, com massa $\mu$.

Essa é uma das demonstrações mais elegantes do poder das simetrias na física:  
**elas transformam um problema aparentemente complexo em algo já conhecido e resolvível.**

# 2. A Derivação Detalhada: Passo a Passo

## 2.1 O Ponto de Partida: O Hamiltoniano de Dois Corpos

Começamos com o Hamiltoniano para duas partículas de massas $m_1$ e $m_2$, interagindo através de um potencial central:

$$
\hat{H} = \frac{\hat{p}_1^2}{2m_1} + \frac{\hat{p}_2^2}{2m_2} + V(|\vec{r}_1 - \vec{r}_2|)
$$

Aqui, $\hat{r}_1, \hat{p}_1$ e $\hat{r}_2, \hat{p}_2$ são os operadores de posição e momento para as partículas 1 e 2, respectivamente.  
Eles obedecem às relações de comutação canônicas:

$$
[\hat{x}_{1i}, \hat{p}_{1j}] = i\hbar \delta_{ij}, \quad
[\hat{x}_{2i}, \hat{p}_{2j}] = i\hbar \delta_{ij}, \quad
[\hat{x}_{1i}, \hat{p}_{2j}] = 0
$$

---

## 2.2 A Mudança de Coordenadas: para o Centro de Massa

Inspirados pela mecânica clássica, definimos as coordenadas do **centro de massa (CM)** e a **posição relativa**:

- **Vetor posição do centro de massa:**

  $$
  \vec{R} = \frac{m_1 \vec{r}_1 + m_2 \vec{r}_2}{m_1 + m_2}
  = \frac{m_1 \vec{r}_1 + m_2 \vec{r}_2}{M}
  $$

  onde $M = m_1 + m_2$ é a **massa total**.

- **Vetor posição relativa:**

  $$
  \vec{r} = \vec{r}_1 - \vec{r}_2
  $$

---

## 2.3 Os Momentos Conjugados e a Transformação Canônica (O Insight da Álgebra)

Toda coordenada tem um momento conjugado.  
Na mecânica quântica, o operador momento é o gerador de translações naquela coordenada:  
$\hat{p}_x = -i\hbar \frac{\partial}{\partial x}$.

Vamos encontrar os operadores de momento $\hat{P}$ e $\hat{p}$ conjugados a $\hat{R}$ e $\hat{r}$.

- **Momento total (conjugado a $\vec{R}$):**

  $$
  \hat{P} = \hat{p}_1 + \hat{p}_2
  $$

- **Momento relativo (conjugado a $\vec{r}$):**

  $$
  \hat{p} = \frac{m_2 \hat{p}_1 - m_1 \hat{p}_2}{m_1 + m_2}
  = \frac{m_2 \hat{p}_1 - m_1 \hat{p}_2}{M}
  $$

A combinação $\frac{\hat{p}_1}{m_1} - \frac{\hat{p}_2}{m_2}$ representa a **velocidade relativa**.

---

### Insight de Álgebra e Mecânica Clássica

A mudança $(\vec{r}_1, \vec{p}_1; \vec{r}_2, \vec{p}_2) \rightarrow (\vec{R}, \vec{P}; \vec{r}, \vec{p})$ é uma **transformação canônica**.

As relações de comutação fundamentais são preservadas:

$$
[\hat{R}_i, \hat{P}_j] = i\hbar \delta_{ij}, \quad
[\hat{r}_i, \hat{p}_j] = i\hbar \delta_{ij}
$$

e, crucialmente, as variáveis do centro de massa comutam com as variáveis relativas:

$$
[\hat{R}_i, \hat{r}_j] = [\hat{R}_i, \hat{p}_j] = [\hat{P}_i, \hat{r}_j] = [\hat{P}_i, \hat{p}_j] = 0
$$

Isto significa que o espaço de Hilbert total se decompõe em um **produto tensorial** de dois subespaços independentes:  
o espaço do CM e o espaço relativo.

---

## 2.4 A Separação do Hamiltoniano (A Mágica da Álgebra)

Vamos reescrever o termo de energia cinética do Hamiltoniano original usando as novas variáveis de momento.

Invertendo as relações, temos:

$$
\hat{p}_1 = \frac{m_1}{M}\hat{P} + \hat{p}, \quad
\hat{p}_2 = \frac{m_2}{M}\hat{P} - \hat{p}
$$

O termo cinético original é:

$$
\frac{\hat{p}_1^2}{2m_1} + \frac{\hat{p}_2^2}{2m_2}
$$

Substituindo:

$$
\frac{1}{2m_1}\left(\frac{m_1}{M}\hat{P} + \hat{p}\right)^2
+ \frac{1}{2m_2}\left(\frac{m_2}{M}\hat{P} - \hat{p}\right)^2
$$

Expandindo os quadrados (lembrando que $\hat{P}$ e $\hat{p}$ comutam):

$$
= \frac{1}{2m_1}\left(\frac{m_1^2}{M^2}\hat{P}^2 + 2\frac{m_1}{M}\hat{P}\cdot\hat{p} + \hat{p}^2\right)
+ \frac{1}{2m_2}\left(\frac{m_2^2}{M^2}\hat{P}^2 - 2\frac{m_2}{M}\hat{P}\cdot\hat{p} + \hat{p}^2\right)
$$

Distribuindo e agrupando:

$$
= \frac{\hat{P}^2}{2M} + \frac{\hat{p}^2}{2\mu}
$$

onde definimos a **massa reduzida**:

$$
\mu = \frac{m_1 m_2}{m_1 + m_2}
$$

Portanto, o Hamiltoniano total se separa elegantemente:

$$
\hat{H} = \frac{\hat{P}^2}{2M} + \left(\frac{\hat{p}^2}{2\mu} + V(r)\right)
$$

Definindo:

- **Hamiltoniano do centro de massa:**
  $$
  \hat{H}_{CM} = \frac{\hat{P}^2}{2M}
  $$

- **Hamiltoniano relativo:**
  $$
  \hat{H}_{rel} = \frac{\hat{p}^2}{2\mu} + V(r)
  $$

Temos:

$$
\hat{H} = \hat{H}_{CM} + \hat{H}_{rel}
$$

---

## 2.5 A Separação da Função de Onda e da Energia

Como $[\hat{H}_{CM}, \hat{H}_{rel}] = 0$, os dois operadores comutam.

Pela Álgebra Linear, isso significa que podemos encontrar autoestados simultâneos e fatorar a função de onda total:

$$
\Psi(\vec{R}, \vec{r}) = \phi(\vec{R}) \psi(\vec{r})
$$

Aplicando o Hamiltoniano total:

$$
\hat{H}\Psi(\vec{R}, \vec{r})
= (\hat{H}_{CM} + \hat{H}_{rel})\phi(\vec{R})\psi(\vec{r})
= (\hat{H}_{CM}\phi(\vec{R}))\psi(\vec{r})
+ \phi(\vec{R})(\hat{H}_{rel}\psi(\vec{r}))
$$

Se $\phi$ e $\psi$ são autoestados de seus respectivos Hamiltonianos:

$$
\hat{H}_{CM}\phi = E_{CM}\phi, \quad
\hat{H}_{rel}\psi = E_{rel}\psi
$$

Então:

$$
\hat{H}\Psi = (E_{CM} + E_{rel})\Psi, \quad
E_{total} = E_{CM} + E_{rel}
$$

Portanto, o problema se decompõe em dois:

1. **Movimento do Centro de Massa:**
   $$
   \frac{\hat{P}^2}{2M}\phi = E_{CM}\phi
   \quad\Rightarrow\quad
   \phi(\vec{R}) = e^{i\vec{P}\cdot\vec{R}/\hbar}
   $$

2. **Movimento Relativo:**
   $$
   \left(\frac{\hat{p}^2}{2\mu} + V(r)\right)\psi = E_{rel}\psi
   $$

Este é **exatamente o problema do potencial central** resolvido na parte anterior da aula.

---

## 2.6 A Separação do Momento Angular

Podemos aplicar o mesmo raciocínio ao momento angular total:

$$
\hat{L} = \hat{L}_1 + \hat{L}_2 = \hat{r}_1 \times \hat{p}_1 + \hat{r}_2 \times \hat{p}_2
$$

Substituindo as novas variáveis e realizando a álgebra, obtemos:

$$
\hat{L} = (\hat{R} \times \hat{P}) + (\hat{r} \times \hat{p})
$$

Definindo:

- **Momento angular do centro de massa:**
  $$
  \hat{L}_{CM} = \hat{R} \times \hat{P}
  $$

- **Momento angular relativo:**
  $$
  \hat{L}_{rel} = \hat{r} \times \hat{p}
  $$

Como $\hat{H}_{CM}$ descreve uma partícula livre, ele comuta com $\hat{L}_{CM}$.  
Como $\hat{H}_{rel}$ descreve uma partícula em um potencial central, ele comuta com $\hat{L}_{rel}$.

Portanto, **os momentos angulares do centro de massa e do movimento relativo são separadamente conservados.**

---

## Conclusão

Através dessa mudança de coordenadas, motivada por simetrias fundamentais, transformamos um problema de dois corpos em 6D em dois problemas independentes e muito mais simples:

1. Uma **partícula livre** (o centro de massa) em 3D;
2. Uma **partícula de massa reduzida** $\mu$ em um potencial central 3D, cujo problema já reduzimos a uma equação radial 1D.

Este é um exemplo notável de como **simetrias e álgebra linear** (operadores que comutam, autoestados simultâneos, decomposição tensorial) permitem resolver sistemas que, à primeira vista, parecem intratáveis.
