# Hydrogen Atom — Radial Equation, Quantization and Hidden SO(4) Symmetry

_Resumo da aula • derivação passo a passo com ênfase nas condições assintóticas, Kummer e degenerescências_


## 0) Setup e notação

- Potencial coulombiano (unidades SI, mas usaremos formas compactas):  
  $$V(r)=-\frac{Ze^2}{r}.$$
- Hamiltoniana:  
  $$\hat H=\frac{\hat{p}^2}{2m}-\frac{Ze^2}{r}.$$
- Em coordenadas esféricas a função de onda separa:  
  $$\Psi(r,\theta,\phi)=R(r)Y_{\ell m}(\theta,\phi).$$
- A equação radial padrão para $$u(r)=r\,R(r)$$ é
  $$
  \boxed{\;\frac{d^2u}{dr^2}+\Big[\frac{2m}{\hbar^2}\big(E- V(r)\big)-\frac{\ell(\ell+1)}{r^2}\Big]u=0\;}\tag{R}
  $$
  (trabalharemos com $$u$$ porque elimina a derivada de primeira ordem).



## 1) Assintótica para $r\to 0$ (pequeno $r$)

Para $r$ muito pequeno, o termo centrífugo $\ell(\ell+1)/r^2$ domina (o termo $$E-V$$ é finito). A Eq. (R) vira
$$
\frac{d^2u}{dr^2}-\frac{\ell(\ell+1)}{r^2}u\simeq 0.
$$
Tenta-se $$u\sim r^s$$ e obtém-se a equação indicial
$$
s(s-1)-\ell(\ell+1)=0\;\Rightarrow\; s=\ell+1\ \text{ou}\ s=-\ell.
$$

- $$s=\ell+1\Rightarrow u\sim r^{\ell+1} \Rightarrow R=u/r\sim r^{\ell}$$ (regular na origem).
- $$s=-\ell\Rightarrow u\sim r^{-\ell}\Rightarrow R\sim r^{-\ell-1}$$ (diverge como $$r\to0$$, não normalizável; às vezes é dito "solução tipo delta" pela divergência forte).

**Conclusão:** escolhemos a solução **regular** $R(r)\propto r^{\ell}$ perto da origem.



## 2) Assintótica para $r\to\infty$ (grande $r$)

Para estados ligados ($$E<0$$), a Eq. (R) para $$r\to\infty$$ reduz a
$$
\frac{d^2u}{dr^2}-\kappa^2 u\simeq 0,\qquad \kappa\equiv\frac{\sqrt{-2mE}}{\hbar}.
$$
Logo $$u\sim e^{-\kappa r}$$ (a solução crescente é descartada). Portanto
$$
R(r)=\frac{u}{r}\sim e^{-\kappa r}/r\;\;\text{e motivamos o ansatz global:}\;\;
\boxed{\;R(r)=(2\kappa r)^{\ell}e^{-\kappa r}\,w(\rho),\quad \rho\equiv 2\kappa r\;}
$$
com $$w(\rho)$$ função regular a determinar.



## 3) Inserindo o ansatz e obtendo a equação de Kummer

Coloque o ansatz em (R) usando $$u=rR=(2\kappa r)^{\ell+1}e^{-\kappa r}w(\rho)$$.  
Depois de derivar cuidadosamente (produto + cadeia) e simplificar, obtém-se:
$$
\boxed{\;\rho\,\frac{d^2w}{d\rho^2}+(2\ell+2-\rho)\frac{dw}{d\rho}+\Big(\frac{Z}{a_0\kappa}-\ell-1\Big)w=0\;}\tag{K}
$$
onde $$a_0\equiv\frac{\hbar^2}{me^2}$$ é o raio de Bohr.

A (K) é **a equação confluentemente hipergeométrica de Kummer**:
$$
x f''+(c-x)f'-af=0,\quad \text{com}\; x=\rho,\; c=2\ell+2,\; a=\ell+1-\frac{Z}{a_0\kappa}.
$$
Suas soluções gerais são as **Confluent Hypergeometric Functions**: $$M(a,c;x)\equiv {}_1F_1(a;c;x)$$ e $$U(a,c;x)$$.



## 4) Condição de normalizabilidade $\Rightarrow$ truncamento da série

Para $$x\to\infty$$, $$M(a,c;x)\sim e^{x}x^{a-c}/\Gamma(a)$$ diverge exponencialmente. A única forma de manter
$$R\propto e^{-\kappa r} M(a,c;2\kappa r)$$ **normalizável** é **forçar o truncamento** da série de $$M$$:
$$
a=-N,\qquad N\in\{0,1,2,\dots\}.
$$
Isso transforma $$M(a,c;x)$$ em um **polinômio** (Laguerre associado):
$$
w(\rho)\propto L_{N}^{2\ell+1}(\rho).
$$

Da definição de $$a$$ segue a condição de quantização
$$
\ell+1-\frac{Z}{a_0\kappa}=-N
\;\;\Rightarrow\;\;
\frac{Z}{a_0\kappa}=N+\ell+1\equiv n\quad (n=1,2,3,\dots).
$$

Como $$\kappa=\sqrt{-2mE}/\hbar$$ e $$a_0=\hbar^2/(me^2)$$, obtemos a energia:
$$
\boxed{\;E_n=-\frac{m e^4 Z^2}{2\hbar^2}\,\frac{1}{n^2}\equiv \frac{E_{\rm gs}}{n^2},\quad E_{\rm gs}= -\tfrac12 m c^2 Z^2\alpha^2\;}
$$
com $$\alpha=\tfrac{e^2}{\hbar c}\approx\tfrac{1}{137}$$.



## 5) Degenerescências: SO(3) vs. o caso do Hidrogênio

- **Apenas SO(3) (rotacional):** a degenerescência esperada vem de $$m\in\{-\ell,\dots,\ell\}$$, de dimensão $$2\ell+1$$.
- **Hidrogênio:** a energia **não** depende de $$\ell$$ (apenas de $$n$$). Para dado $$n$$, $$\ell=0,1,\dots,n-1$$.  
  A degenerescência total do nível $$n$$ é
  $$
  g(n)=\sum_{\ell=0}^{n-1}(2\ell+1)=n^2,
  $$
  que é **maior** do que se esperaria de SO(3). Isto sinaliza uma **simetria dinâmica adicional**.



## 6) Vetor de Laplace–Runge–Lenz (LRL) e órbitas clássicas

Para o potencial $$V\propto-1/r$$ (lei do inverso do quadrado), as órbitas clássicas são **elipses fechadas**, e existe a constante vetorial de movimento (LRL):
$$
\vec A=\frac{1}{2m}(\vec p\times\vec L-\vec L\times\vec p)-Ze^2\,\hat{\vec r}.
$$
No caso quântico **coulombiano**,
$$
[\hat H,\vec A]=0.
$$
> O LRL é **específico** de potenciais do tipo $$-1/r$$ (análogos existem para outros casos, e.g. oscilador harmônico possui simetria dinâmica SU(3)).



## 7) Álgebra SO(4) a partir de $\vec L$ e do LRL

No subespaço de energia ligada ($$E<0$$), define-se o vetor **adimensional**
$$
\vec N\equiv \Big(-\frac{m}{2E}\Big)\vec A\;\;(=\,\vec A/\!\sqrt{-2mE}\,).
$$
Os comutadores tornam-se
$$
[ L_i, L_j ] = i\hbar\,\varepsilon_{ijk} L_k,\qquad
[ N_i, L_j ] = i\hbar\,\varepsilon_{ijk} N_k,\qquad
[ N_i, N_j ] = i\hbar\,\varepsilon_{ijk} L_k,
$$
que fecham a **álgebra so(4)**.

Conveniente introduzir
$$
\vec J_\pm\equiv\tfrac12(\vec L\pm\vec N),
$$
então
$$
[ J_{\pm i}, J_{\pm j}] = i\hbar\,\varepsilon_{ijk} J_{\pm k},\qquad
[ J_{+i}, J_{-j}] = 0,
$$
i.e., $$\text{so}(4)\simeq \text{su}(2)\oplus\text{su}(2).$$

A condição de Casimir dá
$$
J_+^2=J_-^2=\frac{n^2-1}{4}\,\hbar^2\;\Rightarrow\;
j_+=j_- = \frac{n-1}{2}.
$$
Logo a dimensão do irrep é
$$
(2j_++1)(2j_-+1)=n\cdot n=n^2,
$$
**explicando a degenerescência $n^2$ dos níveis de energia do hidrogênio.**



## 8) Forma final das autofunções radiais

Juntando tudo, as soluções radialmente normalizadas (até constante de normalização) são
$$
R_{n\ell}(r)\propto \Big(\frac{2Zr}{na_0}\Big)^{\ell}e^{-\frac{Zr}{na_0}}\,L^{2\ell+1}_{\,n-\ell-1}\!\Big(\frac{2Zr}{na_0}\Big).
$$

---

### TL;DR lógico da quantização
1. **Origem:** manter $$R\sim r^{\ell}$$ (descarta a solução divergente).  
2. **Infinito:** impor decaimento $$e^{-\kappa r}$$.  
3. **Ansatz:** fatorar os comportamentos assintóticos e reduzir a Kummer.  
4. **Normalização:** exigir truncamento $\Rightarrow$ $$a=-N$$, com $$N\in\mathbb{N}_0$$.  
5. **Espectro:** $$n=N+\ell+1\Rightarrow E_n=E_{\rm gs}/n^2$$.  
6. **Degenerescência:** SO(4) via $$\vec L,\vec A$$ $\Rightarrow$ $$n^2$$.
