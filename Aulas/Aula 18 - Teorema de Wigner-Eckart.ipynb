# Aula 18 — Teorema de Wigner–Eckart  
## Operadores Tensorais, Seleção de Regras e Elemento de Matriz Reduzido

Nesta aula estudamos o Teorema de Wigner–Eckart, que fornece uma das ideias mais poderosas da Mecânica Quântica: a separação entre **dependência geométrica** (descrita pelos coeficientes de Clebsch–Gordan) e **dependência dinâmica** (concentrada no elemento reduzido).  

O objetivo pedagógico desta aula é entender:

- por que operadores que transformam como tensores irredutíveis têm propriedades semelhantes às dos estados de momento angular;
- como isso leva a **regras de seleção** automáticas para $m$ e $j$;
- por que os elementos de matriz se fatorizam;
- e como essa fatorização simplifica cálculos físicos.

A intenção conceptual é mostrar que *a estrutura de simetria do operador determina completamente sua ação angular*, e apenas **um único número dinâmico** precisa ser calculado de fato.

## 1. Operadores Tensorais Irredutíveis

Um conjunto de operadores $\{T^{(k)}_q\}$ com $q=-k,\ldots,+k$ forma um **tensor irredutível de ordem $k$** se satisfaz:

### Comutador com $J_z$:
$$
[J_z,\, T^{(k)}_q] = \hbar q\, T^{(k)}_q
$$

**Intenção pedagógica:**  
Esse comutador diz que cada componente $T^{(k)}_q$ "carrega" um valor de projeção $q$ de momento angular, assim como um estado $|k,q\rangle$.  
Ou seja: **o operador se transforma como um objeto de momento angular $k$**.

---

### Comutador com $J_\pm$:
$$
[J_\pm,\, T^{(k)}_q]
= \hbar\sqrt{(k\mp q)(k\pm q+1)}\, T^{(k)}_{q\pm 1}
$$

**Intenção pedagógica:**  
Esse segundo comutador mostra como os componentes do tensor mudam sob operações de escalonamento.  
Isso é crucial para garantir que o conjunto $\{T^{(k)}_q\}$ realmente forma uma representação irredutível de dimensão $2k+1$ do grupo SU(2)/SO(3).

Ele estabelece uma **analogia perfeita** entre operadores tensorais e estados de momento angular.

## 2. Primeira consequência: regra de seleção para $m$

Queremos estudar o elemento de matriz:
$$
\langle \alpha' j' m'| T^{(k)}_q |\alpha j m \rangle
$$

Usando o primeiro comutador:

$$
0 = \langle \alpha' j' m' | [J_z , T^{(k)}_q] - \hbar q\,T^{(k)}_q | \alpha j m \rangle
$$

Expandindo o comutador:
$$
[J_z,T^{(k)}_q] = J_zT^{(k)}_q - T^{(k)}_qJ_z
$$

Aplicando sobre os estados:
$$
\begin{aligned}
0
&= (m'\hbar)\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
- (m\hbar)\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
- q\hbar \langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle \\
&= \hbar (m' - m - q)\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
\end{aligned}
$$

Portanto:
$$
\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle = 0  
\quad \text{a menos que} \quad m' = m + q.
$$

**Intenção pedagógica:**  
Esta demonstração mostra exatamente como a estrutura de simetria rotacional **impõe automaticamente** regras de seleção.  
Antes de qualquer cálculo explícito, já sabemos quando um elemento de matriz é zero.

## 3. A ideia da fatorização

Queremos escrever:
$$
\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
$$

de forma separada em uma parte geométrica e uma parte dinâmica.

O Teorema de Wigner–Eckart afirma que:

$$
\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
= \langle \alpha' j' \| T^{(k)} \| \alpha j \rangle
\;
\langle j m \, k q | j' m' \rangle
$$

onde:

- **o segundo fator** é um **coeficiente de Clebsch–Gordan**, totalmente geométrico;  
- **o primeiro fator** é o **elemento reduzido**, independente de $m$, $m'$ e $q$.

**Intenção pedagógica:**  
O professor quer mostrar que tudo o que depende dos números quânticos magnéticos é *universal*, ou seja, é determinado apenas pela simetria.  
Toda a física específica do sistema está escondida no elemento reduzido.

## 4. Estrutura da demonstração completa

A prova consiste em:

1. Usar o comutador com $J_\pm$:
   $$
   [J_\pm, T^{(k)}_q]
   = \hbar\sqrt{(k\mp q)(k\pm q+1)} T^{(k)}_{q\pm 1}
   $$

2. Avaliar o elemento de matriz:
   $$
   \langle \alpha' j' m'| [J_\pm, T^{(k)}_q] |\alpha j m\rangle
   $$

3. Expandir o comutador:
   $$
   \langle j'm'|J_\pm T - T J_\pm|jm\rangle
   $$

4. Usar as regras de atuação de $J_\pm$:
   $$
   J_\pm |jm\rangle = \hbar \sqrt{(j\mp m)(j\pm m+1)} |j,m\pm1\rangle
   $$

5. Obter uma equação linear recorrente entre três elementos de matriz diferentes.

Essa equação tem exatamente a mesma estrutura da **equação recursiva dos Clebsch–Gordan**.

Portanto, concluímos que a dependência nos índices magnéticos deve ser proporcional a um CG:

$$
\langle \alpha' j' m'|T^{(k)}_q|\alpha j m\rangle
= C\, \langle j m \, k q | j' m' \rangle
$$

onde $C$ não depende de $m$, $m'$, $q$.

Chamamos esse $C$ de **elemento de matriz reduzido**.

## 5. Elemento de matriz reduzido

Denotamos:
$$
\langle \alpha' j' \| T^{(k)} \| \alpha j \rangle = C
$$

Este número contém **toda a física do operador**: carga, massa, estrutura radial, etc.  
O que sobra é puramente geométrico.

**Intenção pedagógica:**  
O professor quer enfatizar que o Teorema permite separar:

- *o que depende da simetria* (CG-coeffs)
- *o que depende da dinâmica* (elemento reduzido)

e isso economiza dezenas de cálculos.

## 6. Exemplos importantes

### 6.1 Operador escalar
Um operador escalar é um tensor de ordem $k=0$:
$$
T^{(0)}_0 = S
$$

O teorema dá:
$$
\langle \alpha' j' m'|S|\alpha j m\rangle
= \delta_{jj'}\delta_{mm'} \langle \alpha' j\|S\|\alpha j\rangle
$$

**Interpretação:**  
Um escalar não muda nem $j$ nem $m$. Apenas multiplica o estado.

---

### 6.2 Operador Vetorial
Um vetor é um tensor de ordem $k=1$:
$$
V_q = T^{(1)}_q
$$

Regras de seleção:
- $\Delta m = m' - m = q = -1,0,+1$
- $\Delta j = j' - j = 0, \pm 1$
- transição $j=0 \to j'=0$ é proibida

Essas regras surgem imediatamente apenas da estrutura tensorial — nenhuma conta explícita é necessária.

---

## 6.3 Teorema de Projeção (caso especial)

Para operadores vetoriais e $j=j'$:

$$
\langle \alpha' j m' | V_q | \alpha j m \rangle
=
\frac{\langle \alpha' j || \mathbf{J}\cdot\mathbf{V} || \alpha j\rangle}
{\hbar^2 j(j+1)}
\langle j m | J_q | j m' \rangle
$$

Esse é conhecido como **Teorema de Projeção**.

## 7. Demonstração do Teorema de Projeção

Escrevemos o operador escalar:
$$
\mathbf{J}\cdot\mathbf{V}
= J_0 V_0 + J_{+1}V_{-1} + J_{-1}V_{+1}
$$

Como é um escalar ($k=0$), o elemento de matriz **não pode depender de $m$**.

Usamos o Teorema de Wigner–Eckart em cada termo:

- todos os três termos são proporcionais ao mesmo elemento reduzido  
- mas com coeficientes CG diferentes  

Portanto:

$$
\langle \alpha' j m'|\mathbf{J}\cdot\mathbf{V}|\alpha j m\rangle
= C_j \langle \alpha' j \| V \| \alpha j \rangle
$$

Para encontrar a constante que relaciona $C_j$ ao elemento reduzido, escolhemos como operador teste $V = J$:

$$
\langle \alpha j m'| J^2 | \alpha j m \rangle
= \hbar^2 j(j+1)\,\delta_{m'm}
$$

Isso fixa a normalização e nos leva ao resultado final:

$$
\langle \alpha j m' | V_q | \alpha j m \rangle
=
\frac{\langle \alpha j || \mathbf{J}\cdot\mathbf{V} || \alpha j\rangle}
{\hbar^2 j(j+1)}\,
\langle j m | J_q | j m' \rangle.
$$

**Intenção pedagógica:**  
O professor mostra como aplicar o Wigner–Eckart em um caso concreto, obtendo um resultado muito usado em física atômica: a projeção de vetores quânticos.
